# Módulos en Python

Hasta ahora siempre hemos utilizado _jupyter notebooks_ para escribir nuestro código Python.

En realidad, los programas Python normalmente se almacenan en ficheros de texto con extensión `.py`. A estos ficheros se les llama _scripts_ Python y en ellos solo puede aparecer código Python (con comentarios). En un _script_ no hay celdas de código o documentación markdown.

Para trabajar con un _script_ Python no podemos usar jupyter; debemos usar al menos:

- un editor simple (preferentemente que coloree los programas Python) para **editar** los _scripts_
- el REPL de Python (`python` o `ipython`) para **ejecutar** los _scripts_

Alternativamente, podemos usar herramientas que permitan tanto ejecutar como editar los _scripts_:

- un editor con soporte para Python, como Visual Studio Code (VS Code)
- un IDE (Integrated Development Environment), como PyCharm

## _Scripts_ Python

Como hemos mencionado, un _script_ Python es un fichero de texto con extensión `.py` que almacena código Python.

### Editando un _script_ Python

Vamos a crear nuestro primer _script_ Python, un programa que lee el nombre del usuario e imprime un saludo.

Utiliza el bloc de notas (`notepad`) para crear el fichero `saludo.py` y copia el siguiente código:

```python
nombre = input("¿Cómo te llamas? ")
print(f"Hola, {nombre}, encantado de conocerte")
```
### Ejecutando un _script_ Python

Un _script_ está pensado para ser ejecutado desde el REPL de Python o desde un IDE. Vamos a ejecutar el _script_ desde el REPL. Para ello, debes:

1. ejecutar `Anaconda Prompt` desde el menú inicio
2. colocarte en el directorio donde esté almacenado `saludo.py` (utiliza la orden `cd`)
3. ejecutar `python saludo.py`

## Usando VS Code para trabajar con _scripts_ Python

Para trabajar con _scripts_ Python es más cómodo utilizar VS Code. Primero necesitarás instalar la extensión `Python`. Para ejecutar el _script_ desde VS Code:

1. Selecciona `File/Open File` y elige el fichero `saludo.py`.<br> Observa que el código Python aparece coloreado y que si colocas el puntero del ratón sobre una función, aparece una ventana emergente con su documentación.
2. En la esquina superior derecha, despliega el menú junto al botón _"play"_ y selecciona `Run Python File in Dedicated Terminal`.

Como ves, usar VS Code es mucho más cómodo que usar el bloc de notas y la terminal. De hecho, si el programa que estás escribiendo es suficientemente grande, VS Code es mucho más cómodo y potente que los _notebooks_.

Para terminar, vamos a escribir un _script_ más grande que define una función `resuelve_ecuacion` para resolver una ecuación de segundo grado. Copia el siguiente código en un fichero `raices.py`:

```python
# script (programa) para calcular las raíces de una ecuación de segundo grado

import math

def resuelve_ecuacion(a,b,c):
    """
    Resuelve una ecuacion de segundo grado a x^2 + bx + c = 0.

    Parameters
    ----------
    a : float
        Coeficiente de segundo grado. No puede ser cero.
    b : float
        Coeficiente de primer grado.
    c : float
        Término independiente.

    Returns
    -------
    Una tupla con las raíces reales.

    Throws
    ------
    ValueError si no hay dos soluciones reales.
    """

    if a == 0:
        print("error: la ecuación no es de segundo grado")
    else:
        discriminante = b ** 2 - 4 * a * c
        if discriminante < 0:
            print("error: las raíces son complejas")
        else:
            raiz_discriminante = math.sqrt(discriminante)
            raices = ((-b + raiz_discriminante) / (2 * a),
                      (-b - raiz_discriminante) / (2 * a))
            return raices


a = float(input("dame el coeficiente a: "))
b = float(input("dame el coeficiente b: "))
c = float(input("dame el coeficiente c: "))
r1, r2 = resuelve_ecuacion(a, b, c)
print("las raíces son" ,r1, r2)
```

Como ves, un _script_ puede definir funciones y usarlas. También puede usar funciones de las bibliotecas de Python (`import math`). Además, el código Python de un _script_ puede (¡y debe!) estar documentado. En esta ocasión hemos utilizado una _docstring_ para documentar la función `resuelve_ecuacion`.

## Gestionando los errores con `raise`

Si utilizas el _script_ anterior para resolver una ecuación que no tenga dos raíces reales, verás que nuestro mensaje de error no es demasiado útil: Python genera su propio mensaje de error mucho más vistoso, luego nuestro mensaje de error pasa desapercibido.

Para evitar esto, vamos a señalar los errores mediante la sentencia `raise`. De esta forma, Python se encarga de comunicar nuestro error de forma más vistosa.

## Módulos Python

Un módulo Python es similar a un _script_ Python, pero no está pensado para ser ejecutado, sino importado. Ya hemos usado varios módulos Python (`math`, `random`, `json`).

Un módulo Python contiene un conjunto de definiciones quer podemos usar desde otros módulos, _scripts_ o _notebooks_.

Podemos convertir el anterior _script_ Python en un módulo simplemente eliminado el código que se ejecuta y dejando solo las definiciones (en nuestro caso, la función `resuelve_ecuacion`).

Borra el contenido anterior del fichero `raices.py` y copia el siguiente código:

```python
# módulo para calcular las raíces de una ecuación de segundo grado

import math


def resuelve_ecuacion(a, b, c):
    """
    Resuelve una ecuacion de segundo grado a x^2 + bx + c = 0.

    Parameters
    ----------
    a : float
        Coeficiente de segundo grado. No puede ser cero.
    b : float
        Coeficiente de primer grado.
    c : float
        Término independiente.

    Returns
    -------
    Una tupla con las raíces reales.

    Throws
    ------
    ValueError si no hay dos soluciones reales.
    """

    if a == 0:
        raise ValueError("la ecuación no es de segundo grado")
    else:
        discriminante = b ** 2 - 4 * a * c
        if discriminante < 0:
            raise ValueError("las raíces son complejas")
        else:
            raiz_discriminante = math.sqrt(discriminante)
            raices = ((-b + raiz_discriminante) / (2 * a),
                      (-b - raiz_discriminante) / (2 * a))
            return raices
```

Observa que `raices.py` ahora no ejecuta ninguna acción, no pide coeficientes ni resuelve una ecuación, simplemente define una función. Si ejecutas `raices.py` desde VS Code (o desde el REPL), verás que no ocurre nada.

## Usando las definiciones de un módulo

Para utilizar las definiciones de un módulo es necesario importarlo mediante la sentencia `import` (es lo que hemos hecho con `math`, `random`, `json`, etc.).

Por ejemplo, podemos usar nuestro módulo `raices.py` desde este mismo _notebook_ (asumiendo que tanto el _notebook_ como `raices.py` están almacenados en el mismo directorio).

Primero debemos importarlo.

````python
import raices
````

Una vez importado, podemos usar las definiciones que ofrece:

````python
raices.resuelve_ecuacion(3,5,1)
````

````python
raices.resuelve_ecuacion(0,5,1)
````

También podemos ecribir el código que se encarga de solicitar los coeficientes por teclado:

````python
a = float(input("dame el coeficiente a: "))
b = float(input("dame el coeficiente b: "))
c = float(input("dame el coeficiente c: "))
r1, r2 = raices.resuelve_ecuacion(a, b, c)
print("las raíces son" ,r1, r2)
````

Finalmente, podemos consultar la documentación de `resuelve_ecuacion`:

````python
help(raices.resuelve_ecuacion)
````

## Módulo y _script_ Python

Al escribir un fichero Python `.py`, no estamos obligados a escoger su papel entre _script_ y módulo. Podemos escribir el fichero `.py` y dejar que sea el usuario quien decida si lo quiere usar como un _script_ (ejecutar las acciones que contiene) o como un módulo (importar las definiciones que contiene).

Añade al final de `raices.py` el siguiente código:

```python
if __name__ == "__main__":
    # esto se ejecuta al usarlo como script, pero no al importarlo
    a = float(input("dame el coeficiente a: "))
    b = float(input("dame el coeficiente b: "))
    c = float(input("dame el coeficiente c: "))
    r1, r2 = resuelve_ecuacion(a, b, c)
    print("las raíces son" ,r1, r2)
```

Observa que lo que estamos haciendo es condicionar la ejecución de las acciones del _script_, que se encargan de preguntar al usuario por los coeficientes de la ecuación y la resuelven.

- Si ejecutamos `raices.py` desde VS Code o desde el REPL, se ejecutarán las acciones del _script_
- Si importamos `raices`, sus definiciones pasan a estar disponibles, pero no se ejecuta el _script. 

## Otras formas de `import`

Hasta ahora hemos utilizado la sentencia `import` de manera muy simple; por ejemplo:

```python
import math
```

Esto nos permite usar todas las definiciones del módulo `math`, siempre que precedamos su nombre con `math.`.

````python
import math
````

````python
math.pi
````

````python
math.sqrt(2)
````

````python
math.cos(math.pi)
````

````python
math.sin(math.pi)
````

````python
help(math)
````

A veces el nombre de un módulo es demasiado largo y resulta molesto escribir su nombre delante de todas sus definiciones. Para evitar esto, podemos utilizar la siguiente forma de `import`:

```python
import nombre_modulo_completo as nombre_modulo_abreviado
```

donde `nombre_modulo_abreviado` es un nombre que nos inventamos nosotros. Para algunos módulos, el nombre abreviado está estandarizado.

````python
import numpy as np
import pandas as pd
import matplotlib as plt
````

A veces un módulo exporta demasiadas definiciones y solo nos interesa utilizar unas pocas. En tal caso, podemos usar una forma de `import` que menciona las definiciones que queremos usar:

```python
from nombre_modulo import nombre_1, nombre_2, ...
```

Una ventaja de esta forma de `import` es que las definiciones importadas (`nombre_1`, `nombre_2`, ...) se utilizan sin precederlas del nombre del módulo.

Por ejemplo, del módulo `math` podemos importar `pi`, `cos`, `sin` y `sqrt`:

````python
from math import pi, cos, sin , sqrt
````

````python
pi
````

````python
sqrt(pi)
````

Finalmente, es posible importar todas las definiciones de un módulo sin mencionarlas:

```python
from nombre_modulo import *
```

Esto permite usar todas las definiciones sin precederlas del nombre del módulo. Sin embargo, esta forma de importación se considera mal estilo y está **desaconsejada**.

````python
from math import *
````
