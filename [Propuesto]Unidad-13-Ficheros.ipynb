Para evitar el uso de rutas completas hasta un fichero podemos **cambiar** la carpeta en que estamos ubicados.

## Los tres pasos para trabajar con un fichero en Python

Para trabajar con un fichero en Python siempre hay que dar tres pasos:

1. **abrir** el fichero
2. **acceder** al fichero (leer o escribir)
3. **cerrar** el fichero

Es muy habitual olvidar el tercer paso (Python no te avisa). Si esto ocurre, nuestro programa es erróneo y podríamos perder datos (es como cerrar un programa sin salvar antes los datos).

## El primer paso: la función `open()`

El primer paso para trabajar con un fichero es abrirlo mediante la función

    open(nombre_de_fichero, modo)
    
donde:
- `nombre_de_fichero` es el nombre del fichero que queremos abrir. Si estamos situados en la carpeta en la que se encuentra el fichero, podemos simplemente facilitar su nombre. Si estamos situados en otra carpeta, debemos facilitar la ruta completa o bien cambiar antes a la carpeta deseada.
- `modo` se refiere al tipo de acceso que queremos realizar sobre el fichero

Por ahora supondremos que `modo` puede ser solo `"r"` (_read_, solo lectura) o `"w"` (_write_, solo escritura).

Vamos a abrir un fichero en modo lectura:

````python
import os

# First, let's navigate to the correct directory where README.txt is located
directorio_correcto = r"c:\Users\xavie\Downloads\VisualStudioProjects\UMA_python_001"
os.chdir(directorio_correcto)
print(f"Directorio actual: {os.getcwd()}")
print(f"Archivos en el directorio: {os.listdir()}\n")

# Now open the file
f = open("README.txt", "r")   # abrir el fichero en modo lectura
````

La variable `f` es de un tipo muy extraño:

````python
print(type(f))
print(f)
````

Lo importante de la variable `f`no es su tipo; lo importante es que a partir de ahora todos los accesos al fichero los haremos a través de métodos que aplicaremos a la variable `f`:

    f.metodo(args)

Ya no utilizaremos el nombre del fichero.

## El tercer paso: el método `close()`

Recuerda que siempre hacen falta tres pasos para trabajar con un fichero Python: abrir, acceder y cerrar.

Para cerrar un fichero usamos el método `close()`. El aspecto de un programa Python simple que accede a un fichero es el siguiente:

```python
f = open('my_file.txt', "r")   # 1. abrir el fichero

...                            # 2. acceder al fichero

f.close()                      # 3. cerrar el fichero
```

````python
f.close()
````

## Lectura de ficheros

Para leer el contenido de un fichero es necesario que este exista previamente. Python soporta varios métodos que permiten leer ficheros de diferentes maneras.

### El método `read()` - todo de golpe

El método `read()` lee todo el contenido del fichero de golpe y lo guarda en una cadena:

````python
# Create a README.txt file with sample content
with open("README.txt", "w") as file:
    file.write("This is a test file for Python file manipulation.\n")
    file.write("It can be used to practice reading, writing, and appending.\n")
    file.write("Line 3: Another line of text.\n")
    file.write("You can also test os.path functions with this file.\n")

with open("README.txt", "r") as f:
    contenido_completo = f.read()
    print(contenido_completo)
````

### Método `readlines()` - todas las líneas de golpe

Observa que con `read()` todo el contenido del fichero se almacena en una sola cadena. El método `readlines()` lee todas las líneas del fichero y las almacena en una lista de líneas:

````python
with open("README.txt", "r") as f:
    lineas = f.readlines()
    print(lineas)   # ['Línea 1\n', 'Línea 2\n', 'Línea 3\n']
````

La ventaja es que esto nos permite iterar sobre el fichero línea a línea:

````python
with open("README.txt", "r") as f:
    for num_linea, linea in enumerate(f, start=1):
        print(f"Línea {num_linea}: {linea.strip()}")
````

### El método `readline()` - línea a línea

Es posible leer solo una línea del fichero mediante el método `readline()`. 

````python
with open("README.txt", "r") as f:
    linea = f.readline()
    while linea:
        print(linea.strip())
        linea = f.readline()
````

Por ejemplo, podemos leer las tres primeras líneas del fichero de la siguiente manera:

````python
with open("README.txt", "r") as f:
    linea1 = f.readline()
    linea2 = f.readline()
    linea3 = f.readline()
    print("Línea 1:", linea1.strip())
    print("Línea 2:", linea2.strip())
    print("Línea 3:", linea3.strip())
````

Después de leer la última línea de un fichero, `readline()` devuelve una cadena vacía (`""`). Esto permite usar un bucle `while` para leer todas las líneas de un fichero:

````python
with open("README.txt", "r") as f:
    linea = f.readline()
    while linea:
        print(linea.rstrip())
        linea = f.readline()
````

### Leyendo ficheros con tildes - parámetro `encoding`

Si un fichero contiene tildes, eñes y otros caracteres _"extraños"_ es posible que no se procese adecuadamente:

````python
# Create a file with accents
with open("archivo_acentos.txt", "w", encoding="utf-8") as f:
    f.write("Ánimo de participación\n")
    f.write("Ejercicio número 2\n")
    f.write("Español, inglés y francés\n")

# Read without specifying encoding (might have issues)
with open("archivo_acentos.txt", "r") as f:
    contenido = f.read()
    print("Sin especificar encoding:")
    print(contenido)
````

Es posible que los acentos del anterior fichero no se hayan procesado bien (depende de la configuración de tu sistema). Esto se debe a la codificación de los caracteres. El fichero del ejemplo utiliza una codificación llamada UTF-8. Podemos indicar la codificación deseada en la función `open()` mediante el parámetro `encoding`: 

````python
# Read with UTF-8 encoding specified
with open("archivo_acentos.txt", "r", encoding="utf-8") as f:
    contenido = f.read()
    print("Especificando encoding='utf-8':")
    print(contenido)
````

## Escritura de ficheros

Para escribir el contenido de un fichero **no** es necesario que este exista previamente. Si el fichero no existe, se crea. Pero **cuidado**, si el fichero ya existe, se **sobreescribe**, por lo que se pierde su contenido previo.

Para escribir en un fichero debemos indicar en la función `open()` que el modo de apertura es escritura, `"w"`. Una vez que hayamos terminado de escribir, debemos cerrar el fichero:

````python
f = open("numeros.txt", "w")
f.write("10\n")
f.write("20\n")
f.write("30\n")
f.close()
print("Fichero numeros.txt creado")
````

### El método `write`

Para escribir contenido en un fichero utilizamos el método `write()`:

````python
with open("escritura.txt", "w") as f:
    f.write("Primera línea\n")
    f.write("Segunda línea\n")
    f.write("Tercera línea\n")

print("Contenido del fichero:")
with open("escritura.txt", "r") as f:
    print(f.read())
````

El siguiente programa escribe una lista de nombres en un fichero. Cada nombre se escribe en una línea:

````python
nombres = ["Alice", "Bob", "Charlie", "Diana", "Eve"]

with open("nombres.txt", "w") as f:
    for nombre in nombres:
        f.write(nombre + "\n")

print("Contenido del fichero nombres.txt:")
with open("nombres.txt", "r") as f:
    print(f.read())
````

Observa que el contenido anterior del fichero ha sido reemplazado por la lista de nombres.

### Añadiendo contenido a un fichero - modo `a`

A veces queremos **añadir** contenido a un fichero que ya existe. No debemos abrirlo en modo escritura, `"w"`, pues perderemos todo su contenido. Para añadir nueva información al final de un fichero que ya existe debemos utilizar el modo `"a"`, de _append_.

Por ejemplo, vamos a añadir algunos nombres al anterior fichero:

````python
nuevos_nombres = ["Frank", "Grace", "Henry"]

with open("nombres.txt", "a") as f:
    for nombre in nuevos_nombres:
        f.write(nombre + "\n")

print("Contenido actualizado del fichero nombres.txt:")
with open("nombres.txt", "r") as f:
    print(f.read())
````

Observa que se ha preservado el contenido del fichero y se han añadido los nuevos nombres al final.

##  Autocerrando los ficheros con `with`

Todos los programas que hemos escrito siguen la misma estructura para cumplir con los tres pasos:

```python
f = open('my_file.txt', "r")   # 1. abrir el fichero

...                            # 2. acceder al fichero

f.close()                      # 3. cerrar el fichero
```

Aunque parezca extraño, es muy habitual olvidar el tercer paso, `close()`. Si esto ocurre, es posible perder los datos que se han _"escrito"_ en el fichero (el motivo es que Python a veces postpone la escritura de los datos hasta ejecutar el `close()`).

Es posible pedirle a Python que cierre automáticamente los ficheros que hemos abierto con la sentencia `with`. La sintaxis en la siguiente:

```python
with open(nombre_fichero, modo) as f:
    código para accecer al fichero
```

El acceso al fichero se realiza en el bloque de código del `with`. Al final de este bloque de código, Python ejecuta automáticamente `f.close()`.

Por ejemplo, podemos leer el contenido de un fichero usando una sentencia `with`:

````python
with open("README.txt", "r") as f:
    contenido = f.read()
    print(contenido)
    print(f"¿Fichero cerrado?: {f.closed}")
````

Observa que tras la sentencia `with` hemos preguntado si el fichero `f` está cerrado (`f.closed`) y la respuesta es afirmativa; Python lo ha cerrado automáticamente.

## Iterando sobre un fichero con un bucle `for`

Al igual que las cadenas, los rangos, las listas, los diccionarios, etc., un fichero puede utilizarse como fuente de datos en un bucle `for`. De hecho, es muy habitual iterar sobre el contenido de un fichero usando un bucle `for`. Si embargo, si no somos cuidadosos es posible que lo hagamos erróneamente.

Observa el siguiente código Python que utiliza un fichero como fuente de datos:

````python
for line in open('my_file.txt', "r"):    # what's wrong with this?
    print(line, end='')
````

Aparentemente funciona correctamente, sin embargo hay un error importante... ¿Sabes qué falla en el código anterior?

La manera correcta de iterar sobre un fichero con un bucle `for` es la siguiente:

````python
with open("README.txt", "r") as f:
    for linea in f:
        print(linea, end='')
# File is automatically closed at the end of the with block
print(f"\n¿Fichero cerrado?: {f.closed}")
````

##  La biblioteca `pickle`

Los anteriores métodos permiten leer y escribir líneas de un fichero. A veces necesitamos guardar datos más complicados que es difícil representar como un conjunto de líneas.

Por ejemplo, imagina que queremos guardar el siguiente objeto Python en un fichero:

````python
x = [ {'Name' : 'David', 'Age' : 25, 'Score': 100}, [1.1, 2.2, 3.3], ("married", False,) ]
x
````

Hemos visto que es posible guardar ciertas cadenas en formato JSON. Sin embargo, a veces es necesario guardar la información de manera más eficiente. La biblioteca `pickle` ofrece una solución.

Primero importamos la biblioteca:

````python
import pickle
````

La función `dump(objeto, fichero_pkl)` vuelca el contenido de un objeto Python en un fichero con formato pickle, un formato diferente de JSON pero más compacto y eficiente. Por ejemplo : 

````python
with open("datos.pkl", "wb") as f:
    pickle.dump(x, f)

print("Objeto guardado en datos.pkl")
````

Una vez almacenado el objeto en un fichero de formato pickle, podemos borrar el objeto mediante la sentencia `del`:

````python
del x
print("Objeto x eliminado")
````

Sin embargo, el objeto no está totalmente perdido. Podemos recuperarlo desde el fichero pickle, mediante la función `load`:

````python
with open("datos.pkl", "rb") as f:
    x = pickle.load(f)

print("Objeto recuperado desde datos.pkl:")
````

## Ejercicio

Escribe un programa que lea una secuencia de enteros por teclado (hasta recibir un 0) y los almacene en un fichero.
Después, debe leer los números del fichero y calcular su suma y su media aritmética.

````python
# Read integers from keyboard until 0
numeros = []
print("Introduce números enteros (0 para terminar):")
while True:
    num = int(input("Número: "))
    if num == 0:
        break
    numeros.append(num)

# Write to file
with open("numeros_ejercicio.txt", "w") as f:
    for num in numeros:
        f.write(str(num) + "\n")

# Read from file and calculate sum and average
with open("numeros_ejercicio.txt", "r") as f:
    numeros_leidos = []
    for linea in f:
        numeros_leidos.append(int(linea.strip()))

suma = sum(numeros_leidos)
media = suma / len(numeros_leidos) if numeros_leidos else 0

print(f"\nNúmeros almacenados: {numeros_leidos}")
print(f"Suma: {suma}")
print(f"Media aritmética: {media:.2f}")
````
