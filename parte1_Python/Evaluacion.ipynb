> Construye un objeto `Persona`. Recibe como parámetros, en el siguiente
> orden, el nombre, la edad y su dni.

~~~ {.py}
▶ def nombre(self)
▶ def edad(self)
▶ def dni(self)
~~~

> Devuelven los valores correspondientes almacenados en el objeto (usa decoradores).

~~~ {.py}
▶ def __str__(self)
~~~

> Devuelve una representación en forma de cadena de la persona, según el formato del
> siguiente ejemplo:
>> ~~~
>> (12345678A, Antonio Pérez, 25 años)
>> ~~~

~~~ {.py}
▶ def __eq__(self,other)
~~~

> Compara la persona actual (self) con la persona recibida como parámetro (other). Dos personas serán iguales sólo si el dni es el mismo, independientemente de mayúsculas y minúsculas.

```python
# clase Persona
class Persona:
    def __init__(self, nombre, edad, dni):
        self.__nombre = nombre
        self.__edad = edad
        self.__dni = dni

    @property
    def nombre(self):
        return self.__nombre

    @property
    def edad(self):
        return self.__edad

    @property
    def dni(self):
        return self.__dni

    def __str__(self):
        return f"({self.__dni}, {self.__nombre}, {self.__edad} años)"

    def __eq__(self, other):
        if not isinstance(other, Persona):
            return False
        return self.__dni.lower() == other.dni.lower()
```

**(2 puntos)** La clase `Diputado` hereda de la clase `Persona`, donde además se permite especificar el partido político al que pertenece, que será un atributo privado de esta clase derivada. La clase ofrecerá el siguiente constructor y métodos públicos:
	
~~~ {.py}
▶ def __init__(self,...)
~~~

> Construye un objeto `Diputado`. Recibe como parámetros, en el siguiente
> orden, el dni, nombre, edad y partido político al que pertenece.

~~~ {.py}
▶ def partido(self)
~~~

> Devuelve la información del partido político del diputado (preferible uso de decorador).

~~~ {.py}
▶ def __str__(self)
~~~

> Devuelve la representación textual del objeto, según el formato del
> siguiente ejemplo:
>> ~~~
>> [Partido A] - (12345678A, Antonio Pérez, 25 años)
>> ~~~


```python
# clase Diputado
class Diputado(Persona):
    # El constructor recibe (dni, nombre, edad, partido) según el enunciado
    def __init__(self, dni, nombre, edad, partido):
        # Llamada al constructor de Persona con (nombre, edad, dni)
        super().__init__(nombre, edad, dni)
        self.__partido = partido   # Atributo privado con doble guion bajo

    @property
    def partido(self):
        return self.__partido

    def __str__(self):
        return f"[{self.__partido}] - ({self.dni}, {self.nombre}, {self.edad} años)"
```

**(3.75 puntos)** Defina la clase `Congreso`,que represente el congreso de los diputados en España. Para ello, esta clase contendrá con atributo privado una lista de diputados, que serán objetos de la clase Diputado definida anteriormente. 

~~~ {.py}

> Construye un objeto `Congreso` vacío (sin diputados).

~~~ {.py}
▶ def obtener_diputados(self)
~~~

> Devuelve la lista de diputados del congreso (usar decorador).

~~~ {.py}
▶ def agregar_diputado(...)
~~~

> Crea un nuevo objeto `Diputado` con el dni, nombre, edad y partido político recibidos como parámetros (el objeto se crea en el método). Si el diputado está en la lista se muestra un mensaje por pantalla, en caso contrario se añade. 

~~~ {.py}
▶ def __str__(self)
~~~

> Devuelve la representación textual del objeto, según el formato del
> siguiente ejemplo (sin considerar los saltos de línea):
> ~~~
> {[Partido Socialista] - (11111111A, Pedro Sánchez, 54 años)
>  [Partido Popular] - (22222222F, Alberto Núñez Feijóo, 64 años)
>  [Sumar] - (33333333C, Yolanda Díaz, 54 años)
>  [VOX] - (77777777T, Santiago Abascal, 50 años)}
> ~~~

~~~ {.py}
▶ def obtener_diccionario(self)
~~~

> Recorre la lista de diputados (bucle for) y guarda en un diccionario el nombre de cada diputado como clave y su edad como valor. Devuelve dicho diccionario.

~~~ {.py}
▶ def guardar_congreso(...)
~~~

> Recibe como parámetro el nombre del archivo y guarda la lista de diputados del Congreso en un archivo utilizando la librería Pickle.



```python
# Escribe el código de la clase Congreso

import pickle

class Congreso:
    def __init__(self):
        self.__diputados = []

    @property
    def obtener_diputados(self):
        return self.__diputados

    def agregar_diputado(self, dni, nombre, edad, partido):
        nuevo = Diputado(dni, nombre, edad, partido)
        # Verificar si ya existe (por DNI)
        for d in self.__diputados:
            if d == nuevo:  # usa __eq__ de Persona (compara DNI)
                print(f"El diputado con DNI {dni} ya está en la lista. No se añade.")
                return
        self.__diputados.append(nuevo)

    def __str__(self):
        if not self.__diputados:
            return "{}"
        representacion = "\n ".join(str(d) for d in self.__diputados)
        return f"{{{representacion}}}"

    def obtener_diccionario(self):
        dic = {}
        for d in self.__diputados:
            dic[d.nombre] = d.edad
        return dic

    def guardar_congreso(self, nombre_archivo):
        with open(nombre_archivo, 'wb') as f:
            pickle.dump(self.__diputados, f)
```
