## 1. Cálculo del índice de masa corporal

El índice de masa corporal (IMC) se puede calcular como:

$$
\mathrm{IMC} = \frac{m}{h^2}
$$

donde $m$ es la masa en kilogramos y $h$ es la estatura en metros. Este índice permite clasificar a las personas en bajo peso, peso normal, sobrepeso u obesidad según la siguiente tabla:

| IMC($kg/m^2$) |    Categoría   |
| ------------- | -------------- |
| Bajo peso     |   $\lt$ 18,5   |
| Normal        |  18,5 - 24.99  |
| Sobrepeso     |    25 - 29.99  |
| Obeso         |    30 $\leq$   |

Escribe un programa Python que dadas una masa $m$ y una estatura $h$ escriba un mensaje indicando a qué categoría corresponde esa persona. Por ejemplo, para $m= 52$ y $h=1.63$ debe escribir:

```
Tu IMC es: 19.57168128269788
Tu categoría de IMC es: Normal
```

````python
# Solicitar masa y estatura al usuario
m = float(input("Introduce tu masa en kilogramos (kg): "))
h = float(input("Introduce tu estatura en metros (m): "))

# Calcular el IMC
imc = m / (h ** 2)

# Clasificar según la tabla
if imc < 18.5:
    categoria = "Bajo peso"
elif imc < 25:
    categoria = "Normal"
elif imc < 30:
    categoria = "Sobrepeso"
else:
    categoria = "Obeso"

# Mostrar resultados
print(f"{m}kg {h}cm --->>")
print(f"Tu IMC es: {imc}")
print(f"Tu categoría de IMC es: {categoria}")
````

## 2. Edad canina

Se suele decir que un año canino equivale a siete años humanos. Sin embargo, esta regla de conversión no tiene en cuenta que los perros alcanzan la edad adulta en aproximadamente dos años. En consecuencia, algunas personas creen que es mejor contar cada uno de los dos primeros años caninos como 10,5 años humanos, y luego contar cada año canino adicional como cuatro años humanos.

Escribe un programa que lea una edad canina y la traduzca a humana usando ambas reglas de conversión de años caninos a años humanos.

````python
# Read dog age in years (can be a decimal)
dog_age = float(input("Introduce la edad canina en años: "))

# Rule 1: 1 dog year = 7 human years
human_rule1 = dog_age * 7

# Rule 2: first two years = 10.5 human years each, then 4 human years per additional year
if dog_age <= 2:
    human_rule2 = dog_age * 10.5
else:
    human_rule2 = 2 * 10.5 + (dog_age - 2) * 4

# Display results
print(f"Edad humana (regla 1 año = 7 años): {human_rule1} años")
print(f"Edad humana (regla de maduración): {human_rule2} años")
````

## 4. Raíces de una ecuación de segundo grado

Dada una ecuación de segundo grado:

$$
    ax^2 + bx + c = 0
$$

se pueden calcular sus raíces por radicales:

$$
x_{1, 2} = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

Escribe un programa Python para calcular las raíces de una ecuación de segundo grado.
Para calcular la raíz cuadrada puedes usar la función `sqrt` del módulo `math`. Si las raíces son complejas, se debe mostrar un mensaje de error.

Por ejemplo, las raíces de $x^2 -5x + 6 = 0$ son $x_1 = 3$ y $x_2 = 2$

````python
import math

# Read coefficients
a = float(input("Introduce el coeficiente a: "))
b = float(input("Introduce el coeficiente b: "))
c = float(input("Introduce el coeficiente c: "))

# Case a == 0 (linear equation)
if a == 0:
    if b == 0:
        if c == 0:
            print("La ecuación tiene infinitas soluciones.")
        else:
            print("La ecuación no tiene solución.")
    else:
        x = -c / b
        print(f"Ecuación lineal. Raíz única: x = {x}")
else:
    discriminant = b**2 - 4*a*c

    if discriminant < 0:
        # Complex roots
        real_part = -b / (2*a)
        imag_part = math.sqrt(-discriminant) / (2*a)
        print(f"Las raíces son complejas: x1 = {real_part} + {imag_part}i , x2 = {real_part} - {imag_part}i")
    elif discriminant == 0:
        x = -b / (2*a)
        print(f"Raíz doble: x = {x}")
    else:
        sqrt_disc = math.sqrt(discriminant)
        x1 = (-b + sqrt_disc) / (2*a)
        x2 = (-b - sqrt_disc) / (2*a)
        print(f"Las raíces son: x1 = {x1} y x2 = {x2}")
````

## 5. Estaciones del año

Un año se divide en cuatro estaciones según la siguiente tabla:

| Estación  | Primer día    |
|-----------|---------------|
| Primavera | 21 marzo      |
| Verano    | 21 junio      |
| Otoño     | 23 septiembre |
| Invierno  | 22 diciembre  |

Escribe un programa que lea un día y un nombre de mes y responda a qué estación pertenece.

````python
# Read day and month from user
day = int(input("Introduce el día: "))
month = input("Introduce el nombre del mes (en minúsculas): ")

# Dictionary to map month names to numbers
months = {
    "enero": 1, "febrero": 2, "marzo": 3, "abril": 4,
    "mayo": 5, "junio": 6, "julio": 7, "agosto": 8,
    "septiembre": 9, "octubre": 10, "noviembre": 11, "diciembre": 12
}

# Convert month name to number
month_num = months.get(month.lower())

# Check if month is valid
if month_num is None:
    print("Mes no válido")
else:
    # Determine the season based on date
    if (month_num == 3 and day >= 21) or (month_num == 4) or (month_num == 5) or (month_num == 6 and day < 21):
        season = "Primavera"
    elif (month_num == 6 and day >= 21) or (month_num == 7) or (month_num == 8) or (month_num == 9 and day < 23):
        season = "Verano"
    elif (month_num == 9 and day >= 23) or (month_num == 10) or (month_num == 11) or (month_num == 12 and day < 22):
        season = "Otoño"
    else:
        season = "Invierno"

    # Output the result
    print(f"La estación es: {season}")
````
