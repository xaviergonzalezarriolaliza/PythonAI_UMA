# Estructuras de Datos en Python

Una estructura de datos es una colección de datos llamados elementos. Ya hemos estudiado varias estructuras de datos proporcionadas por Python (listas, diccionarios...)

Todas las estructuras de datos deben definir operaciones para:

- crear la estructura vacía
- añadir un dato a la estructura
- consultar un dato de la estructura
- eliminar un dato de la estructura
- saber el número de datos de la estructura
- saber si una estructura está vacía

Para una lista `l` estas operaciones son:

- crear una lista vacía: `l = []`
- añadir: `l.append(x)`
- consultar (indexación): `l[i]`
- eliminar: `l.pop(i)`
- número de elementos: `len(l)`
- está vacía: `len(l) == 0` o `l == []`

En este tema vamos a estudiar cómo implementar nuestras propias estructuras de datos usando clases Python (`class`).

## Pilas

Piensa en la forma en que manipulamos una pila de platos cuando estamos recogiendo o poniendo la mesa:

- al poner la mesa, retiramos el plato limpio que está encima de la pila de platos
- al recoger la mesa, depositamos un plato sucio sobre la pila de platos

Una **pila** es una estructura de datos que se comporta, presisamente, como una pila de platos:

- al insertar un dato, lo colocamos encima de los demás datos
- al retirar un dato, retiramos el que está arriba del todo

Por ese motivo decimos que las pilas son estructuras LIFO (_Last In, First Out_), el dato más recientemente insertado es el dato que se elimina o consulta.

Puedes pensar en una pila como una colección de datos dispuestos verticalmente (como la pila de platos):

     8  <-- cima
     1
     2
     5  <-- base

el dato que está arriba se denomina **cima** de la pila; el dato que está debajo se llama **base** de la pila.

Para una pila `p`, las operaciones típicas de las estructuras de datos son:

- crear una pila vacía: `p = Stack()`
- añadir: `p.push(x)`, añade `x` sobre la cima de la pila `p`
- consultar: `p.top()`, devuelve (sin eliminar) el dato que está en la cima de la pila `p`
- eliminar: `p.pop()`, elimina (sin devolver) la cima de la pila `p`
- número de elementos: `p.size()`, devuelve la longitud de la pila `p`
- está vacía: `p.is_empty()`, indica si la pila `p` está vacía

Observa que todas las operaciones se realizan sobre el dato que ocupa la **cima** de la pila:

                                     3
     cima --> 8                      8                    8                                           4
              1    == push(3) ==>    1    == pop() ==>    1    == pop() ==>    1    == push(4) ==>    1
              2                      2                    2                    2                      2
     base --> 5                      5                    5                    5                      5

Una vez implementada la pila, la importaremos en el _notebook_ y la usaremos:

```python
import Stack

p = Stack.Stack()

p.push(5)
p.push(2)
p.push(1)
p.push(8)
p.push(3)

p.pop()
p.pop()

print(p.top(), p.size())

p2 = Stack.Stack()
p2.push(100)

print(p == p)
print(p == p2)
print(p == "")

print(p)
```

### Ejercicio: Validación de paréntesis balanceados

Se desea comprobar si una cadena de caracteres que contiene paréntesis, llaves y corchetes está correctamente balanceada.

Los símbolos considerados son:

- Paréntesis: ( )
- Corchetes: [ ]
- Llaves: { }

#### Definición de cadena balanceada

Una cadena está correctamente balanceada si:

1. Cada símbolo de apertura tiene su correspondiente símbolo de cierre del mismo tipo.
2. Los símbolos están correctamente anidados.
3. El orden de cierre respeta el orden inverso de apertura (estructura tipo pila).
4. No quedan símbolos de apertura sin cerrar al final.

#### Entrada

Una cadena de caracteres que puede contener:
- paréntesis ( )
- corchetes [ ]
- llaves { }
- otros caracteres (que pueden ignorarse si se desea)

#### Salida

Un valor booleano:
- True si la cadena está balanceada
- False en caso contrario

#### Ejemplos

Entrada: ([]{})
Salida: True

Entrada: ([)]
Salida: False

Entrada: ((()))
Salida: True

Entrada: (((
Salida: False

## Colas

Piensa la forma en la que organizamos una cola en la caja de un supermercado:

- cuando llega un nuevo cliente, se coloca al final de la cola y pasa a ser el último de la cola
- cuando le toca al siguiente cliente, el primero de la cola es atendido y abandona la cola

Una cola es una estructura de datos que se comporta, precisamente, como una cola en un supermercado:

- al insertar un dato, lo colocamos al final de la cola
- al retirar un dato, retiramos el primero de la cola

Por ese motivo decimos que las pilas son estructuras FIFO (_First In, First Out_), el dato que se insertó primero es el dato que se elimina o consulta.

Puedes pensar en una cola como una colección de datos dispuestos horizontalmente:

     primero     último
     |           |
     8   1   2   5

como en cualquier cola, un dato es el **primero** y otro dato es el **último**.

Para una pila `q`, las operaciones típicas de las estructuras de datos son:

- crear una cola vacía: `q = Queue()`
- añadir: `q.enqueue(x)`, añade `x` al final de la cola `q`
- consultar: `q.first()`, devuelve (sin eliminar) el primero de la cola `q` 
- eliminar: `q.dequeue()`, elimina (sin devolver) el primero de la cola `q`
- número de elementos: `q.size()`, devuelve la longitud de la cola `q`
- está vacía: `q.is_empty()`, indica si la cola `q` está vacía

Observa que algunas operaciones afectan al **primero** de la cola y otras al **último**:

     primero     último
     |           |
     8   1   2   5    == enqueue(3) ==>    8   1   2   5   3    == dequeue() ==>    1   2   5   3    == dequeue() ==>    2   5   3


**Ejercicio:**

Vamos a implementar estas operaciones en el módulo `Queue.py`. También añadiremos operaciones para destruir una cola (`del`), obtener una cadena que represente la cola (`str`) y saber si dos colas son iguales (`eq`).

Una vez implementada la cola, la importaremos en el _notebook_ y la usaremos:

```python
import Queue

q = Queue.Queue()

q.enqueue(8)
q.enqueue(1)
q.enqueue(2)
q.enqueue(5)
q.enqueue(3)

q.dequeue()
q.dequeue()

print(q)
```

### Problema de Josephus

1. Se elige un número entero `k`.
2. Se comienza desde la persona número 1.
3. Se cuenta en sentido circular (1, 2, 3, ..., n, 1, 2, ...).
4. Cada vez que se llega a la k-ésima persona, esta es eliminada del círculo.
5. El proceso continúa desde la siguiente persona, reiniciando la cuenta.
6. Se repite el proceso hasta que solo queda una persona.

#### Objetivo

Dado `n` y `k`, determinar qué posición (número de persona) es la última en sobrevivir.

#### Entrada

- Un entero `n` (número de personas).
- Un entero `k` (salto de eliminación).

#### Salida

- Una lista con el orden de ejecución de las personas ejecutadas.
- Un entero que representa la posición de la última persona restante.

#### Ejemplo

n = 7, k = 3

Proceso de eliminación:
- Se elimina la persona 3
- Luego la 6
- Luego la 2
- Luego la 7
- Luego la 5
- Luego la 1

Salida:
[6,2,7,5,1]
