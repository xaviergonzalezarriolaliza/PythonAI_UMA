## Necesidad de las funciones

Imagina que en un programa tenemos que calcular las medias aritméticas de pesos y estaturas de un número de personas:

````python
pesos = [82, 81, 54, 48, 102, 77]
estaturas = [1.78, 1.85, 1.61, 1.58, 1.9, 1.83]

suma_peso = 0
for peso in pesos:
    suma_peso += peso
peso_medio = suma_peso / len(pesos)

print("el peso medio es", peso_medio)

suma_estatura = 0
for estatura in estaturas:
    suma_estatura += estatura
estatura_media = suma_estatura / len(estaturas)

print("la estatura media es", estatura_media)
````

Observa que el código para calcular la media aritmética aparece dos veces. Las apariciones son virtualmente idénticas, solo cambia el nombre de las variables, pero el código es esencialmente el mismo y hace el mismo cálculo.

De la misma forma que los bucles permiten escribir una sola vez el código que se repite de forma consecutiva, las funciones permiten escribir una sola vez el código que se repite, pero ni de forma consecutiva ni se aplica a los mismos datos.

Para evitar repetir el código que hace virtualmente lo mismo con diferentes datos, necesitamos una herramienta que nos permita:
- escribir el código que realiza un cálculo una sola vez
- usar ese código cuando sea necesario
- facilitar en cada uso los datos que debe utilizar para el cálculo (`pesos`, `estaturas`)
- permitir al código que nos informe del resultado del cálculo (`peso_medio`, `estatura_media`)

Esa herramienta son las **funciones**. Observa cómo la función `input()`:
- contiene el código necesario para leer de teclado
- la podemos usar cuando sea necesario
- al usarla, podemos facilitar el mensaje que se debe mostrar por pantalla (`input("dime tu nombre:")`)
- nos devuelve un resultado con la cadena que se ha tecleado (`nombre = input("dime tu nombre:")`)

Similarmente, nos gustaría escribir el cálculo de la media aritmética una sola vez y usarlo cuando lo necesitemos; es decir, nos gustaría contar con una función `media_aritmetica()`, facilitarle los datos apropiados para realizar el cálculo y obtener el resultado:

```python
peso_medio = media_aritmetica(pesos)
estatura_media = media_aritmetica(estaturas)
```

## Definición básica de funciones

Al definir una función, lo que hacemos es dar un nombre al código que queremos usar en varias ocasiones. La sintaxis de la definición es:
```python
def nombre_funcion():
    bloque_de_código_de_la_función  # cuerpo de la función
```

Observa que, como las sentencias de control, una función introduce un bloque de código. A ese bloque se le llama **cuerpo** de la función.

Por ejemplo, vamos a definir una función `asteriscos()` que imprima una línea con 80 asteriscos:

````python
def asteriscos():
    print(80 * '*')
````

Para usar una función basta usar su nombre seguido de un par de paréntesis. A esto se le denomina **llamar** o **invocar** a una función:

````python
asteriscos()
````

### Pasando un argumento

La función `asteriscos()` siempre realiza el mismo cálculo, imprime 80 asteriscos. Imagina que queremos que cada vez que la usemos podamos decidir cuántos asteriscos queremos imprimir. Para ello, debemos añadir en la definición de la función un **parámetro**:

```python
def nombre_funcion(parámetro):
    bloque_de_código_de_la_función
```

En la definición anterior, `parámetro` es una variable que podremos utilizar dentro de la función; es la forma que tenemos que comunicarle a la función los datos que queremos emplear en el cálculo. Obviamente, el `bloque_de_código_de_la_función`puede utilizar la variable `parámetro`.

Por ejemplo, vamos a definir una función `asteriscos(num_asteriscos)` que imprima la cantidad de asteriscos que necesitemos:

````python
def asteriscos(num_asteriscos):
    print(num_asteriscos * '*')
````

````python
asteriscos(5)
````

Al invocar la función `asteriscos()` debemos facilitar entre paréntesis el número de asteriscos que queremos imprimir. A esto se le llama **pasar un argumento** a la función; es la forma que tenemos de comunicar a una función los datos con los que queremos que haga el cálculo.

El valor que pasamos a la función recibe el nombre de **argumento** y se utiliza para **inicializar** el **parámetro** de la función. Esto permite ajustar el tamaño de las líneas de la cabecera:
```

hola python
***********
```

````python
n = len("hola python")
asteriscos(n)
print("hola python")
asteriscos(n)
````

### Pasando varios argumentos

Una función puede recibir más de un argumento. Basta indicar los parámetros separados por comas en la definición de la función:

```python
def nombre_funcion(parámetro_1, parámetro_2, ..., parámetro_n):
    bloque_de_código_de_la_función
```

Al invocar a la función habrá que facilitar un argumento (valor inicial) para cada parámetro.

Por ejemplo, vamos a definir una función `imprime_linea(n_veces, caracter)` que imprima `n_veces` un `caracter`:

````python
def imprime_linea(n_veces, char):
    for i in range(n_veces):
        print(char, end='')
    
imprime_linea(10, '*-__')
````

Obviamente, el cuerpo de la función puede estar compuesto por más de una línea de código y puede contener invocaciones a otra función. Vamos a definir una función `cabecera(titulo)` que imprima la cadena `titulo` en una cabecera de asteriscos, de manera que `cabecera("Hola Python")` imprimirá:

````python
def cabecera(titulo):
    n_asteriscos = len(titulo) + 4
    asteriscos(n_asteriscos)
    print("* " + titulo + " *")
    asteriscos(n_asteriscos)
cabecera("Hello Python")
````

## Funciones y procedimientos

Como hemos comentado, hay funciones como `len()` o `input()` que devuelven un resultado; mientras que otras funciones como `print()` no devuelven ningún resultado. A las funciones que no devuelven resultados se les llama **procedimientos**.

La distinción entre función y procedimiento es similar a la distinción entre expresión y sentencia:

- las funciones devuelven un valor, al igual que las expresiones
- los procedimientos no devuelven un valor; tienen un efecto, al igual que las sentencias

De hecho, una llamada a función es una expresión y una llamada a procedimiento es una sentencia.

Sin embargo, en Python la distinción entre función y procedimiento no es tan clara como en otros lenguajes. En realidad, los procedimientos en Python sí devuelven algo:

Aunque en los _notebooks_ no se imprime nada, realmente se ha devuelto algo que podemos imprimir con `print()`:

El valor devuelto es `None`, algo así como nada. ¿De qué tipo es `None`?

````python
x=None
````

El literal `None` representa el **único valor** del tipo `NoneType`.

Podemos crear variables de tipo `NoneType`:

````python
x=NoneType
````

Con ello indicamos que aún no conocemos un valor adecuado para la variable; pero la variable en sí existe.

Es habitual usar `None` para indicar que un cálculo no puede completarse satisfactoriamente. Por ejemplo, supongamos que queremos definir
una función para calcular el recíproco de un número. En principio podemos pensar en una definición directa:

````python
def reciproco(n):
    return 1 / n if n != 0 else None

print(reciproco(0))
````

El problema es que esta definición falla si se pide calcular el recíproco de 0:

````python
1/0
````

Para evitar ese error, podemos devolver `None` cuando el recíproco no exista:

````python

print(reciproco(0))
````

```python
def nombre_funcion(*argumentos):  # número variable de argumentos

```
El parámetro `argumentos` es una **tupla**: podemos acceder a cada argumento mediante indexación y podemos iterar con un bucle `for`. Además, podemos saber el número de argumentos mediante la función `len()`.

Vamos a definir una función `varios_argumentos(args)` que pueda recibir un número variables de argumentos. La función debe imprimir el número de argumentos que ha recibido, así como el valor y el tipo de cada uno de ellos.

````python
def varios_argumentos(*args):
    num = len(args)
    print(f"Se recibieron {num} argumento(s):")
    for i, valor in enumerate(args, start=1):
        print(f"  Argumento {i}: valor = {valor}, tipo = {type(valor)}")

varios_argumentos(10, "hola", 3.14, [1, 2])
````

```python
    bloque_de_código_de_la_función
```

Si un parámetro tiene valor por defecto, no es necesario facilitar su valor inicial en la invocación; es decir, no es necesario poner un argumento en la invocación.

## Paso de argumentos por `keyword`

Hasta el momento hemos utilizado paso de parámetros **posicional**. El primer argumento de la invocación inicializa al primer parámetro de la definición, y así sucesivamente. En Python, es posible pasar los argumentos usando el **nombre de los parámetros**. Esto es algo que ya hemos hecho con la función `print()`, usando `sep` y `end`:

````python
print("hola", "mundo", sep = ', ', end = ". ")
print("¿cómo va", "todo", end = "?", sep = ' ')
````

Observa que estamos facilitando a la función `print()` los valores iniciales de sus parámetros `end` y `sep`. Para ello, inicializamos los parámetros de forma explícita en la invocación. Esto permite que la posición de un argumento no sea la de su correspondiente parámetro. El argumento se asocia a su parámetro por nombre, no por posición.

Recuerda la última definición de la función `imprime_linea(n_veces, caracter)`, que inicializaba ambos parámetros por defecto:

````python
def imprime_linea(n_veces=1, char="x"):
    for i in range(n_veces):
        print(char, end='')
````

Finalmente, podemos **mezclar** una invocación **posicional** y por `keyword`, pero los primeros argumentos deben ser posicionales y se corresponden con los primeros parámetros de la función:

````python
def mult_y_div(*args, m, d):
    """
    Multiplica cada valor en args por m y lo divide por d.
    *args: argumentos posicionales (números).
    m, d: argumentos obligatorios por keyword.
    """
    # Utilizamos una lista por comprensión para procesar los datos de forma eficiente
    return [(val * m) / d for val in args]
````

Mezclar invocación **posicional** y por `keyword` es muy útil para invocar funciones que toman un número variable de argumentos. Es algo que ya hemos hecho con la función `print()`:

```python

```

Observa que las cadenas `"hola"` y `"mundo"` son varios argumentos que corresponden a un solo parámetro (`*args`); mientras que `sep` y `end` son argumentos pasados por `keyword`

````python
# Ejemplo de invocación mezclando tipos de argumentos:
# 10, 20, 30 son capturados por *args
# m y d se pasan explícitamente por nombre
valores = [10, 20, 30]
resultado = mult_y_div(10, 20, 30, m=10, d=2)

print("Valores originales:", valores)
print("Factores: m=10, d=2")
print("Resultado:", resultado)
````

## Ámbito de una función

Consideremos el siguiente programa Python:

```python
a = 100

def f():
    a = 500
    print("dentro de f():", a)

print("antes de invocar a f():", a)
f()
print("después de invocar a f():", a)
```

Al ejecutarlo, ¿cuál será el valor final de `a`?

````python
def circulo(radio, pi=3.1416):
    area = pi * radio ** 2
    circunferencia = 2 * pi * radio
    return (area, circunferencia)

print
````

Observa que la asignación `a = 500` no tiene efecto fuera de la función. La variable `a` sigue valiendo `100` después de invocar a `f()`. Esto se debe a que las funciones Python introducen un **ámbito**; es decir, la variable `a` que aparece en el bloque de la función `f()` es una variable **nueva**, **local** a `f()`. La variable `a` que aparece fuera de la función `f()` es una variable **global**. Ambas variables son **distintas**, las asignaciones a una no afectan a la otra.

Cuando una función Python asigna un valor a una variable por primera vez, se crea una variable nueva, local a la función.

```python
a = 100        # variable global

def f():
    a = 500    # variable local (distinta de la global)
    print(a)

print(a)
f()
print(a)
```

Una consecuencia de lo anterior es que una función Python no puede modificar las variables que recibe como parámetros.

Por ejemplo, si definimos una función `incrementa(x)` que incrementa en una unidad el argumento recibido, veremos que, en realidad, el argumento no es modificado:

````python
n=10

def incrementa(x):
    # Al realizar esta asignación, Python crea una variable local 'x'
    # que apunta a un NUEVO objeto entero (el resultado de x + 1).
    x = x + 1
    print(f"  [Dentro de incrementa] El valor local de x es: {x}")
    return x

print(f"Antes de la función: n = {n}")

print(f"Increment(n)={incrementa(n)}")


print(f"Después de la función: n = {n}")

# El resultado será:
# Antes de la función: n = 10
#   [Dentro de incrementa] El valor local de x es: 11
# Después de la función: n = 10
````

### Asignación a tipos mutables e indexables

Los tipos mutables e indexables (como las listas) son una excepción a la anterior regla.

Esto se debe a que si usamos indexación para modificar una posición, el dato debe existir para que el código no falle. Por ello, Python asume que el dato es externo a la función:

````python
# Definimos una lista en el ámbito global
mi_coleccion = [1, 2, 3]

def modificar_por_indice():
    # Nota: No estamos declarando 'global mi_coleccion'
    # Al usar [0], Python busca 'mi_coleccion' fuera de la función,
    # la encuentra y modifica su primer elemento.
    mi_coleccion[0] = "Cambiado"

def intentar_reasignar():
    # Esto SÍ fallaría o crearía una variable local si intentamos 
    # usarla antes, porque la asignación directa '=' sin índice 
    # define un nuevo nombre en el ámbito local.
    mi_coleccion = [4, 5, 6] # Esto NO modificaría la global
    pass

print(f"Estado inicial: {mi_coleccion}")

modificar_por_indice()
print(f"Después de modificar_por_indice: {mi_coleccion}")

# Resultado:
# El cambio persiste porque hemos mutado el objeto, no reasignado la variable.
````

En cualquier caso, se trata de una práctica **desaconsejada**: una función no debería modificar un dato que no recibe como parámetro.

De hecho, esta excepción se aplica también cuando pasamos el dato mutable e indexable como parámetro, pero en esta ocasión hacemos explícito que la función tiene acceso al dato que va a modificar:

````python
def duplicar_elementos(lista):
    """
    Modifica la lista original multiplicando cada elemento por 2.
    Al ser un objeto mutable, los cambios persisten fuera de la función
    porque operamos sobre la referencia recibida.
    """
    for i in range(len(lista)):
        lista[i] *= 2

# Definimos la lista original
mis_numeros = [1, 2, 3, 4, 5]

print(f"Antes de la función: {mis_numeros}")

# Al pasar 'mis_numeros', hacemos explícito que la función trabajará con este dato
duplicar_elementos(mis_numeros)

print(f"Después de la función: {mis_numeros}")
````

### Declaración de variable global

A veces necesitamos que una función Python modifique una variable global; es decir, que las asignaciones del cuerpo de la función afecten a la variable global. En tal caso, debemos incluir en la función una declaración de variable global; para indicar que no se debe crear una variable local nuevo, sino utilizar la variable global:

```python
def nombre_funcion(parámetro_1, parámetro_2, ..., parámetro_n):
    global variable_global                                       # declaración de variable global
    bloque_de_código_de_la_función
    return resultado_1, resultado_2, ..., resultado_m
```

En general, se **desaconseja** el uso de declaraciones de variables globales en Python.

Definiremos una función `incrementa()`, pero esta vez incrementará un variable global:

````python
# Declaramos una variable en el ámbito global
contador = 0

def incrementa():
    # Indicamos que queremos usar la variable 'contador' definida fuera
    global contador
    contador = contador + 1
    print(f"  [Dentro de la función] contador es ahora: {contador}")

print(f"Antes de llamar a la función: {contador}")

incrementa()
incrementa()
incrementa()

print(f"Después de las llamadas: {contador}")
````

### Modificación de colecciones mutables en funciones Python

Recuerda que una asignación en un bloque de función crea una nueva variable local (excepto si la variable asignada ha sido declarada `global`). Por otro lado, las colecciones mutables se pueden modificar sin utilizar la sentencia de asignación. Ciertos métodos (por ejemplo, `append` sobre listas) pueden modificar una colección.

Observa la diferencia entre las siguientes definiciones de la función `inserta(ls, v)`, que inserta un valor `v` al final de una lista `ls`.

La primera definición utiliza una sentencia de asignación (sin indexación):

````python
def inserta_con_asignacion(ls, v):
    # Intento fallido de modificar la lista externa
    ls = ls + [v]
    print(f"  Dentro de la función (asignación): {ls}")

print("1. Probando inserta_con_asignacion:")
inserta_con_asignacion(mi_lista, 30)
print(f"Resultado fuera: {mi_lista}") # Sigue siendo [10, 20]
````

La segunda definición utiliza el método de listas `append()`:

````python
def inserta_con_append(ls, v):
    # Modificación exitosa del objeto original
    ls.append(v)
    print(f"  Dentro de la función (append): {ls}")

print("\n2. Probando inserta_con_append:")
inserta_con_append(mi_lista, 30)
print(f"Resultado fuera: {mi_lista}") # Ahora es [10, 20, 30]
````

## Solución de la media aritmética mediante funciones

## Señalando errores en funciones

La función `media_aritmetica` no está definida si la lista está vacía. Para señalar ese error, podemos utilizar un `print()` o devolver `None`.

En cualquier caso, se trata de una mala solución. Observa que Python señala ese tipo de errores de una manera mucho más clara:
- detiene la ejecución del programa
- muestra un mensaje de error, indicando el tipo de error (`TypeError`, `IndexError`, ...)
- señala el código Python que ha provocado el error

Por ejemplo, observa la forma en que Python señala los siguientes errores:

````python
raise ValueError
````

Es evidente que la forma de señalar los errores de Python es preferible a mostrar un mensaje con `print()`, que podría pasar desapercibido entre una maraña de mensajes, o a devolver un `None` que podríamos no usar.

Para señalar un error como la hace Python, basta usar la sentencia `raise`, cuya sintaxis es:

```python
   raise tipo_de_error("mensaje de error")
```

donde `tipo_de_error` puede ser `TypeError`, `IndexError`, etc. Nosotros usaremos siempre `ValueError`.

Podemos modificar la función `media_aritmetica()` para que señale los errores como lo hace Python:

````python
def media_aritmetica(lista):
    # Comprobamos si la lista está vacía para evitar errores de división
    if len(lista) == 0:
        raise ValueError("La lista no puede estar vacía")
    
    suma = 0
    for numero in lista:
        suma += numero
        
    return suma / len(lista)

# Ejemplo de cómo capturar este error:
try:
    mi_lista = []
    resultado = media_aritmetica(mi_lista)
    print("La media es:", resultado)
except ValueError as e:
    print(f"Error detectado: {e}")
````
