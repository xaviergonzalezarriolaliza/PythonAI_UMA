# Unidad 8 - Sentencias Iterativas - I

Hasta el momento todos nuestros programas Python han utilizado dos **estructuras de control**: la **secuencia** y la **selección**.

En el caso de la selección, hemos introducido varias sentencias Python:

- selección simple (`if`)
- selección alternativa (`if else`)
- selección múltiple (`if elif else`)

Estas sentencias, combinadas con el **anidamiento** de control, nos permiten diseñar la lógica de nuestros programas usando **condiciones** para controlar el **flujo** de ejecución.

Siempre debemos utilizar estas estructuras de control para escribir programas correctos, simples y eficientes. Por ejemplo, el siguiente programa permite clasificar las notas de un alumnos realizando solo dos comprobaciones por nota:

```python
if 0 <= nota < 7:
    if nota < 5:
        calificacion = "SUSPENSO"
    else:
        calificacion = "APROBADO"
else:
    if nota < 9:
        calificacion = "NOTABLE"
    else:
        calificacion = "SOBRESALIENTE"
```

Sin embargo, hay situaciones en las que la secuencia y la selección no resultan suficientes para escribir nuestros programas. En esta unidad introduciremos una nueva estructura de control: la **iteración**.

## Repitiendo n veces un código

Los problemas anteriores se resolverían muy fácilmente si pudiéramos construir bucles en Python; es decir, indicar que queremos repetir cierto código un número de veces.

Para esto podemos usar la sentencia `for`. Su sintaxis es:

```python
for i in range(número_de_veces):
    bloque_de_código_a_repetir
```

donde:
- El `bloque_de_código_a_repetir`es un **bloque** con las sentencias a repetir, se le llama **cuerpo** del bucle
- El entero `numero_de_veces` indica cuántas veces se repetirá el código

**Nota:** el nombre `i` que aparece entre `for` e `in` puede reemplazar por cualquier otro nombre válido en Python. Por ahora lo ignoraremos.

- imprimir los cuadrados del 1 al 10 (observa que el bloque tiene dos sentencias)

````python
for i in range(100):
    print("no se habla en clase ", i)
````

````python

n = 10
for i in range(1, n+1):
    print("Avanzo una calle")

print("Giro a la izquierda")
````

## Variable de control del bucle `for`

El nombre que aparece entre `for` e `in` es una variable, se le llama la **variable de control** del bucle.

Si un bucle `for` se repite $n$ veces, la variable de control toma automáticamente los valores $0,1,2,3\dots n-1$. La variable de control se puede consultar dentro del cuerpo del bucle.

Por ejemplo, para el siguiente bucle `for`, la variable de control toma sucesivamente los valores $0,1,2,3,4$:

````python
for i in range(0, 5, 1): # Start, stop, step
    print(i)
````

````python
for n in range(0, 5, 1): # Start, stop, step
    print(n)
````

Puedes elegir cualquier otro nombre para la variable de control, su nombre no afecta al bucle:

````python
for variablecontrol in range(0, 5, 1): # Start, stop, step
    print(variablecontrol)
````

### Variables de control y banderas

En la siguiente figura, el ciclista tiene que dar cinco vueltas a una rotonda hasta antes de pasar por una verja. Para llevar la cuenta de las vueltas que ha dado, el ciclista recoge una bandera justo antes de completar una vuelta.



Podemos representar el comportamiento del ciclista con un bucle `for`. La variable de control contará el número de banderas que transporta el ciclista justo antes de recoger la siguiente bandera:

````python
for banderas in range(0, 5, 1): # Start, stop, step
    print(banderas, "banderas")
````

### Variable de control anónima

A veces el cuerpo del bucle no necesita acceder a la variable de control. En tales ocasiones, podemos utilizar la variable de control _anónima_, `_`:

```python
for _ in range(número_de_veces):
    bloque_de_código_a_repetir
```

````python
for _ in range(10):
    print("En clase no se habla")
````

## Repitiendo para una secuencia aritmética de valores

Los ejemplos anteriores repiten el cuerpo del bucle $n$ veces: si no necesitamos saber en qué repetición nos encontramos, podemos usar la variable anónima; de lo contrario utilizamos una variable de control cuyos valores serán $0,1,2,3\dots n-1$.

A veces necesitamos repetir un código de forma de la variable de control no vaya de $0$ a $n-1$, sino que tome valores equidistantes dentro de un rango dado:
- los números pares entre 0 y 20
- los números impares entre 1 y 100
- los múltiplos de 7 entre 0 y 100
- los enteros entre -5 y 5

Para tales casos debemos indicar en `range` el valor  inicial, el valor final (sin incluir) y la diferencia entre los números visitados (recuerda, son equidistantes):

```python
   range(inicio, fin, diferencia)
```

La variable control tomará los valores en el intervalo $[\textrm{inicio}, \textrm{inicio} + \textrm{diferencia}, \textrm{inicio} + 2 \times \textrm{diferencia}  \dots \textrm{fin} )$. Recuerda, el valor $\textrm{fin}$ no está incluido.

````python
for i in range(0, 20, 2):
    print(i)
````

````python
for i in range(1, 100, 2):
    print(i)
````

````python
for i in range(7, 100, 7):
    print(i)
````

## Repitiendo para una secuencia arbitraria de valores

A veces queremos repetir un código para una secuencia de enteros que no son equidistantes (no forman una secuencia aritmética). En tal caso, no podemos utilizar `range`; debemos facilitar todos los valores que queremos visitar encerrados entre corchetes y separados por comas:

```python
for i in [x0, x1, x2, ..., xn]:
    bloque_de_código_a_repetir
```

El código se repetirá $n+1$ veces, y la variable de control tomará los valores $x_0, x_1, x_2\dots x_n$.

Además, los valores $x_i$ no tienen que ser enteros (`int`); pueden ser flotantes (`float`), cadenas (`str`), booleanos (`bool`), o cualquier otro tipo Python. De hecho, ni siquiera es necesario que todos los valores de la secuencia tengan el mismo tipo.

**Nota:** La expresión `[x0, x1, x2, ..., xn]` es una lista; las estudiaremos en el siguiente capítulo.

## Repitiendo para los caracteres de una cadena

A veces queremos repetir un código para **cada carácter** de una cadena de caracteres (`srt`). En tal caso, basta facilitar a Python la cadena cuyos caracteres queremos visitar:

```python
for caracter in cadena:
    bloque_de_código_a_repetir
```
En este caso, el bucle se repetirá $n =$`len(cadena)` veces y la variable de control `caracter` tomará como valores cada uno los caracteres de la cadena, del primero al último.

## (Casi) Toda la verdad sobre el bucle `for`

Hemos visto que el bucle `for` puede adoptar diferentes formas:
- `for i in range(ini, fin, dif)`
- `for i in [x0, x1, x2,,,xn]`
- `for i in "cadena"`
- `for i in variable`

donde la `variable` debe ser de un tipo **iterable**.

La sintaxis general es:
```python
for variable_de_control in fuente_de_datos: # la variable_de_control toma cada uno de los datos de la fuente
    bloque_de_código_a_repetir
```
Observa que la sentencia es la misma independientemente de la `fuente_de_datos` que usemos (`range`, lista, etc.). Al bucle `for` le da igual qué tipo de fuente fuente de datos usemos siempre que sea correcta; se limita a visitar uno a uno los datos de la fuente de datos.

## Operadores y funciones sobre rangos

Hasta ahora hemos utilizado la función `range()` exclusivamente como fuente de datos para el bucle `for`.

En realidad, la función `range()` devuelve un valor de tipo `range` que representa de forma compacta un rango de valores.

Los rangos soportan los siguientes operadores:

|   Operador     |    Significado    |
|----------------|-------------------|
| `[]`           |  indexación       |
| `in`, `not in` |  pertenencia      |


y las siguientes funciones:

|   Función      |    Significado    |
|----------------|-------------------|
| `len`          |  longitud         |
| `reversed`     |  inverso          |

Observa que hemos aplicado todos estos operadores y funciones (excepto `reversed`) a las cadenas.

## El método `format`

Ya hemos visto que podemos imprimir en pantalla cualquier información con la función `print()`. Por ejemplo, podemos usar `print()` para imprimir el código de identificación de un estudiante y su nota final:

````python
codigo = 4537
nota = 19 / 3
print("codigo de la estudiante:", codigo, "nota final:", nota)

codigo = 684
nota = 28 / 3
print("codigo de la estudiante:", codigo, "nota final:", nota)
````

Aunque `print()` suele ser suficiente, a veces queremos controlar el formato con el que se imprime la información. Por ejemplo, podemos suponer que todos los números de identificación tienen exactamente tres dígitos, y no tiene sentido imprimir tantos decimales en una nota final.

El método `format` permite especificar el **formato **(tamaño y forma) de los datos que queremos imprimir:

```python
   "cadena_formato".format(exp1, exp2, exp3, ...)
```

la cadena `"cadena formato"` es una cadena Python que contiene **especificaciones** de formato para cada valor `expi` que queramos imprimir. Estas especificaciones de formato se escriben entre llaves tienen la siguiente sintaxis:

| Especificación de formato | Significado                               |
|---------------------------|-------------------------------------------|
| `:nd`                   | entero con anchura `n`                    |
| `:n.df`                  | flotante con anchura `n` y `d` decimales  |
