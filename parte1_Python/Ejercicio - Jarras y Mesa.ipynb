Ejercicio 1. Jarras
--------------------------------

El objetivo de este ejercicio es crear una clase `Jarra`. Utilizaremos esta clase
`Jarra` para *simular* algunas de las acciones que podemos realizar
con una jarra.

Nuestras jarras van a poder contener cierta cantidad de agua. Así,
cada jarra tiene una determinada capacidad (en litros) que será la
misma (constante) durante la vida de la jarra, y su valor será
especificado en el momento de la construcción del objeto. En un
momento determinado una jarra dispondrá de una cantidad de agua que
podrá variar en el tiempo. Las acciones que podremos realizar sobre
una jarra son:

- Llenar la jarra por completo desde un grifo.

- Vaciar la jarra completamente.

- Llenar la jarra con el agua que contiene otra jarra (bien hasta que
  la jarra receptora quede llena o hasta que la jarra que volcamos se
  vacíe por completo).

- Acceder a su representación textual.

Por ejemplo: Disponemos de dos jarras **A** y **B** de capacidades 7 y
4 litros respectivamente. Podemos llenar la jarra **A** (no podemos
echar menos del total de la jarra porque no sabríamos a ciencia cierta
cuánta agua tendría). Luego volcar **A** sobre **B** (no cabe todo,
por lo que en **A** quedan 3 litros y **B** está llena).  Ahora vaciar
**B**. Después volver a volcar **A** sobre **B**. En esta situación,
**A** está vacía y **B** tiene 3 litros.

Hay que definir la clase `Jarra` con los métodos necesarios para
realizar las operaciones que acabamos de describir. Por lo tanto, la
clase `Jarra` contiene dos variables de instancia, de tal forma que la
variable de instancia `capacidad` determina
la capacidad del objeto `Jarra`, mientras que la variable de instancia
`contenido` determina la cantidad de líquido contenido en un momento
dado en el objeto `Jarra`, considerando que el contenido de la jarra
podrá variar a lo largo del tiempo. Además, la clase `Jarra` define
los siguientes constructores y métodos públicos que permitirán
manipular los objetos de la clase `Jarra`:

~~~ {.py}
▶ def __init__(self,capacidad)
~~~

> Construye un nuevo objeto `Jarra` con la *capacidad* que se recibe
> como parámetro, y el *contenido* de la jarra vacío.

~~~ {.py}
▶ @property
▶ def capacidad(self)
~~~

> Devuelve la *capacidad* del objeto jarra.

~~~ {.py}
▶ @property
▶ def contenido(self)
~~~

> Devuelve el *contenido* actual del objeto jarra.

~~~ {.py}
▶ def llena(self)
~~~

> Llena el *contenido* del objeto jarra al máximo de su capacidad.

~~~ {.py}
▶ def vacia(self)
~~~

> Vacía el *contenido* del objeto jarra completamente.

~~~ {.py}
▶ def llenaDesde(self, other)
~~~

> Llena el *contenido* de la jarra actual (receptora) con el
> *contenido* de la jarra que se recibe como parámetro (emisora), bien
> hasta que la jarra receptora quede llena o hasta que la jarra
> emisora se vacíe por completo.

~~~ {.py}
▶ def __str__(self)
~~~

> Devuelve un cadena de caracteres con la representación textual del objeto jarra
  en el formato `J(`*cap,cnt*`)`.

Para probar nuestra nueva clase:

- Se crearán dos jarras, una (`jarraA`) con
  capacidad para 7 litros y otra (`jarraB`) para 4 litros. Una vez
  creadas hemos de realizar las siguientes operaciones: llenar
  `jarraA`, mostrar por pantalla ambas jarras, volcar `jarraA` sobre
  `jarraB`, mostrar por pantalla ambas jarras, vaciar `jarraB`,
  mostrar por pantalla ambas jarras, volcar `jarraA` sobre `jarraB`,
  mostrar por pantalla ambas jarras.

```python
# X4\/¡ G
# Ejercicio 1 - Classe Jarra
class Jarra:

    def __init__(self, capacidad):
        self._capacidad = capacidad
        self._contenido = 0

    @property
    def capacidad(self):
        return self._capacidad

    @property
    def contenido(self):
        return self._contenido

    def llena(self):
        self._contenido = self._capacidad

    def vacia(self):
        self._contenido = 0

    def llenaDesde(self, other):
        espacio_libre = self._capacidad - self._contenido
        transferencia = min(espacio_libre, other._contenido)
        self._contenido += transferencia
        other._contenido -= transferencia

    def __str__(self):
        return f"J({self._capacidad},{self._contenido})"


if __name__ == "__main__":
    jarraA = Jarra(7)
    jarraB = Jarra(4)

    print("Estado inicial:")
    print(f"jarraA = {jarraA}, jarraB = {jarraB}")

    jarraA.llena()
    print("\nDespués de llenar jarraA:")
    print(f"jarraA = {jarraA}, jarraB = {jarraB}")

    jarraB.llenaDesde(jarraA)
    print("\nDespués de volcar jarraA sobre jarraB:")
    print(f"jarraA = {jarraA}, jarraB = {jarraB}")

    jarraB.vacia()
    print("\nDespués de vaciar jarraB:")
    print(f"jarraA = {jarraA}, jarraB = {jarraB}")

    jarraB.llenaDesde(jarraA)
    print("\nDespués de volcar jarraA sobre jarraB nuevamente:")
    print(f"jarraA = {jarraA}, jarraB = {jarraB}")
```

siguientes acciones con las jarras de la mesa, especificando cada

~~~ {.py}

> Devuelve una cadena de caracteres con la representación textual del objeto mesa
  en el formato `M(J(`*cap,cnt*`),J(`*cap,cnt*`))`.

Para probar nuestra nueva clase:

- Se creará una mesa con dos jarras, una con
  capacidad para 5 litros y otra para 7.  Una vez creada, hemos de
  realizar las operaciones necesarias para dejar en una de las jarras
  exactamente un único litro de agua.

```python
# X4\/¡ G

# Colores ANSI
BLUE = '\033[94m'
CYAN = '\033[96m'
GREEN = '\033[92m'
YELLOW = '\033[93m'
RESET = '\033[0m'

class Jarra:
    def __init__(self, capacidad):
        self._capacidad = capacidad
        self._contenido = 0

    @property
    def capacidad(self):
        return self._capacidad

    @property
    def contenido(self):
        return self._contenido

    def llena(self):
        self._contenido = self._capacidad

    def vacia(self):
        self._contenido = 0

    def llenaDesde(self, other):
        espacio = self._capacidad - self._contenido
        transferencia = min(espacio, other._contenido)
        self._contenido += transferencia
        other._contenido -= transferencia

    def __str__(self):
        return f"J({self._capacidad},{self._contenido})"


class Mesa:
    def __init__(self, capacidad_jarra_1, capacidad_jarra_2):
        self.jarra1 = Jarra(capacidad_jarra_1)
        self.jarra2 = Jarra(capacidad_jarra_2)

    def capacidad(self, num_jarra):
        return self.jarra1.capacidad if num_jarra == 1 else self.jarra2.capacidad

    def contenido(self, num_jarra):
        return self.jarra1.contenido if num_jarra == 1 else self.jarra2.contenido

    def llena(self, num_jarra):
        if num_jarra == 1:
            self.jarra1.llena()
        else:
            self.jarra2.llena()

    def vacia(self, num_jarra):
        if num_jarra == 1:
            self.jarra1.vacia()
        else:
            self.jarra2.vacia()

    def llenaDesde(self, num_jarra):
        # num_jarra es la emisora (vierte sobre la otra)
        if num_jarra == 1:
            self.jarra2.llenaDesde(self.jarra1)
        else:
            self.jarra1.llenaDesde(self.jarra2)

    def __str__(self):
        return f"M({self.jarra1},{self.jarra2})"


# Programa principal: obtener 1 litro en la jarra de 5 (jarra1)
if __name__ == "__main__":
    mesa = Mesa(5, 7)
    print("Estado inicial:", mesa)

    # Secuencia eficiente (8 movimientos)
    mesa.llena(1)            # 1. Llenar jarra1 (5,0)
    print("1. Llena jarra1:", mesa)

    mesa.llenaDesde(1)       # 2. Volcar jarra1 sobre jarra2 -> (0,5)
    print("2. Vierte jarra1->jarra2:", mesa)

    mesa.llena(1)            # 3. Llenar jarra1 (5,5)
    print("3. Llena jarra1:", mesa)

    mesa.llenaDesde(1)       # 4. Volcar jarra1 sobre jarra2 -> (3,7)
    print("4. Vierte jarra1->jarra2:", mesa)

    mesa.vacia(2)            # 5. Vaciar jarra2 (3,0)
    print("5. Vacía jarra2:", mesa)

    mesa.llenaDesde(1)       # 6. Volcar jarra1 sobre jarra2 -> (0,3)
    print("6. Vierte jarra1->jarra2:", mesa)

    mesa.llena(1)            # 7. Llenar jarra1 (5,3)
    print("7. Llena jarra1:", mesa)

    mesa.llenaDesde(1)       # 8. Volcar jarra1 sobre jarra2 -> (1,7)
    print("8. Vierte jarra1->jarra2:", mesa)

    # Resultado final con arte ASCII y colores
    litros = mesa.contenido(1)  # 1 litro en la jarra1

    # Arte ASCII de una jarra con agua
    jar_art = f"""
  {CYAN}  _____
 /     \\
|  {BLUE}~~~{CYAN}  |
|  {BLUE}~~~{CYAN}  |
|  {BLUE}~~~{CYAN}  |
 \\_____/{RESET}
"""

    print(f"\n{jar_art}")
    print(f"{GREEN}✨ ¡En la jarra1 (capacidad 5) hay {litros} litro de agua! ✨{RESET}")
```
