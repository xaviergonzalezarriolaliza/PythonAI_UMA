## Composición

La composición es un concepto clave en la programación orientada a objetos (POO). Se refiere a la relación entre dos clases donde una clase contiene una **instancia de otra clase** como uno de sus miembros o atributos. Esta relación se basa en la idea de que un objeto está **compuesto** por otros objetos. La composición no se limita a una sola clase; puedes tener múltiples clases que se componen entre sí formando una estructura más compleja y jerárquica.

### Ejercicio 0. Media, máximo y mínimo

Usando iteración sobre listas, escribe un programa que calcule el máximo, el mínimo y la media aritmética de una lista de números. La respuesta se debe almacenar en una tupla con tres elementos. Por ejemplo, para la lista `[13, 8, 9, 45, 1, 0, 0, 6, 12]` se debe obtener la tupla `(45, 0, 10.44)`.

```python
# Lista de números
numeros = [13, 8, 9, 45, 1, 0, 0, 6, 12]

# Inicializar variables
maximo = numeros[0]
minimo = numeros[0]
suma = 0

# Recorrer la lista
for num in numeros:
    if num > maximo:
        maximo = num
    if num < minimo:
        minimo = num
    suma += num

# Calcular media y redondear
media = round(suma / len(numeros), 2)

# Crear tupla
resultado = (maximo, minimo, media)

# Mostrar resultado
print(resultado)
```

### Ejercicio 1 
Definir la clase `EstadisticasLista`, que permite trabajar con una lista de números enteros introducidos por el usuario y calcular estadísticas básicas sobre dicha lista. La clase debe tener un atributo `numeros`, que almacena la lista de números introducida por el usuario, y tres atributos adicionales: `maximo`, `minimo` y `media`, que almacenan respectivamente el valor máximo, mínimo y la media aritmética de la lista.

Debe incluir los siguientes métodos:

- `introducir_numeros()`: solicita al usuario una lista de números enteros separados por comas (por ejemplo, `"1,2,3"`) y almacena los valores como enteros en el atributo `numeros`.

- `computar()`: calcula y almacena en los atributos correspondientes el valor máximo, el valor mínimo y la media aritmética de la lista. No realiza validaciones; se asume que la lista ya ha sido introducida correctamente.

- `obtener_maximo()`: devuelve el valor almacenado en el atributo `maximo`.

- `obtener_minimo()`: devuelve el valor almacenado en el atributo `minimo`.

- `obtener_media()`: devuelve el valor almacenado en el atributo `media`.

Crear una instancia de la clase `EstadisticasLista`, introducir una lista de números usando el método `introducir_numeros()`, llamar al método `computar()` para calcular las estadísticas, y luego mostrar el máximo, mínimo y media usando los métodos correspondientes.


```python
# Definicioclass EstadisticasLista:

class EstadisticasLista:    
    # Clase para almacenar una lista de números enteros y calcular
    # el máximo, mínimo y media aritmética.

    def __init__(self):
        # Inicializa los atributos con valores por defecto.
        self.numeros = []
        self.maximo = None
        self.minimo = None
        self.media = None

    def introducir_numeros(self):
        # Solicita al usuario una lista de números enteros separados por comas.
        # Ejemplo de entrada: "1,2,3"
        # Almacena la lista de enteros en el atributo `numeros`.
        entrada = input("Introduce una lista de números enteros separados por comas: ")
        # Dividir por comas, eliminar espacios y convertir a entero
        self.numeros = [int(x.strip()) for x in entrada.split(",")]

    def computar(self):
        # Calcula el máximo, mínimo y media aritmética de la lista `numeros`.
        # Se asume que `numeros` ya está correctamente inicializado.
        if not self.numeros:
            raise ValueError("La lista de números está vacía. No se pueden calcular estadísticas.")

        self.maximo = max(self.numeros)
        self.minimo = min(self.numeros)
        self.media = sum(self.numeros) / len(self.numeros)

    def obtener_maximo(self):
        # Devuelve el valor del atributo `maximo`
        return self.maximo

    def obtener_minimo(self):
        # Devuelve el valor del atributo `minimo`
        return self.minimo

    def obtener_media(self):
        # Devuelve el valor del atributo `media`
        return self.media
```

```python
# Uso de la clase

if __name__ == "__main__":
    # Crear una instancia de la clase
    stats = EstadisticasLista()

    # Solicitar los números al usuario
    stats.introducir_numeros()

    # Calcular las estadísticas
    stats.computar()

    # Mostrar los resultados usando los métodos getter
    print(f"Máximo: {stats.obtener_maximo()}")
    print(f"Mínimo: {stats.obtener_minimo()}")
    # Mostrar la media con dos decimales
    print(f"Media: {stats.obtener_media():.2f}")
```

### Ejercicio 2
Definir las clases `Punto` y `Rectangulo`. La clase `Punto` representa un punto en un plano cartesiano y tiene los atributos `x` e `y` que representan las coordenadas del punto. La clase `Rectangulo` representa un rectángulo en el plano cartesiano y tiene los atributos `punto1` y `punto2`, que son instancias de la clase `Punto` que representan los puntos opuestos del rectángulo.

La clase `Punto` debe tener un método `mostrar_coordenadas()` que imprima las coordenadas del punto en el formato `(x, y)`.

La clase `Rectangulo` debe tener un método `calcular_area()` que calcule y devuelva el área del rectángulo utilizando la fórmula `base * altura`, donde `base` es la diferencia en valor absoluto entre las coordenadas x de los puntos y `altura` es la diferencia en valor absoluto entre las coordenadas y de los puntos.

Además, la clase `Rectangulo` debe tener un método `mostrar_informacion()` que muestre las coordenadas de los dos puntos que forman el rectángulo y el área del mismo.

Crear instancias de las clases `Punto` y `Rectangulo`, pasando los valores adecuados para los puntos, y llamar al método `mostrar_informacion()` de la clase `Rectangulo` para mostrar la información del rectángulo.

```python
class Punto:
    # Representa un punto en el plano cartesiano
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def mostrar_coordenadas(self):
        # Imprime las coordenadas del punto en formato (x, y)
        print(f"({self.x}, {self.y})")
```

```python
# Crear dos puntos
p1 = Punto(-200, -333)
p2 = Punto(80, -666)
```

```python
class Rectangulo:
    # Representa un rectángulo definido por dos puntos opuestos
    
    def __init__(self, punto1, punto2):
        # Inicializa el rectángulo con dos instancias de Punto
        self.punto1 = punto1
        self.punto2 = punto2
    
    def calcular_area(self):
        # Calcula el área como base * altura (valores absolutos)
        base = abs(self.punto1.x - self.punto2.x)
        altura = abs(self.punto1.y - self.punto2.y)
        return base * altura
    
    def perimetro(self):
        base = abs(self.punto1.x - self.punto2.x)
        altura = abs(self.punto1.y - self.punto2.y)
        return 2*base+2*altura
    
    def mostrar_informacion(self):
        # Muestra coordenadas de ambos puntos y el área
        print("Coordenadas del punto 1:", end=" ")
        self.punto1.mostrar_coordenadas()
        print("Coordenadas del punto 2:", end=" ")
        self.punto2.mostrar_coordenadas()
        print(f"Área del rectángulo: {self.calcular_area()}")
        print(f"Perímetro: {self.perimetro()}")
```

```python
# Crear rectángulo con esos dos puntos opuestos
rect = Rectangulo(p1, p2)

# Mostrar la información
rect.mostrar_informacion()
```

### Ejercicio 3

Definir las clases `Universidad`, `Facultad` y `Estudiante` en Python. La clase `Universidad` representa una universidad y tiene un atributo `nombre`. La clase `Facultad` representa una facultad en la universidad y tiene un atributo `nombre` y una lista de objetos `Estudiante` que representan a los estudiantes matriculados en la facultad.

La clase `Estudiante` representa a un estudiante y tiene los atributos `nombre` y `codigo`. La clase `Estudiante` también tiene un método `mostrar_informacion()` que muestra por pantalla el nombre y el código del estudiante.

La clase `Facultad` tiene un método `agregar_estudiante()` que recibe un objeto `Estudiante` y lo agrega a la lista de estudiantes de la facultad. La lista será un atributo privado y contará un metodo `estudiantes()` que devolverá dicha lista (utiliza el decorador *@property*).

Crear instancias de las clases `Universidad`, `Facultad` y al menos dos objetos `Estudiante`. Luego, agregar los estudiantes a la facultad y llamar al método `mostrar_informacion()` de cada estudiante para mostrar su información.

**Nota:** La relación entre `Facultad` y `Estudiante` es de composición, ya que una facultad contiene estudiantes y la existencia de los estudiantes depende de la existencia de la facultad a la que pertenecen.

```python
class Universidad:
    # Representa una universidad con un nombre
    def __init__(self, nombre):
        self.nombre = nombre
```

```python
class Facultad:
    # Representa una facultad con nombre y lista privada de estudiantes
    def __init__(self, nombre):
        self.nombre = nombre
        self.__estudiantes = []  # Atributo privado
    
    def agregar_estudiante(self, estudiante):
        # Agrega un objeto Estudiante a la lista privada
        self.__estudiantes.append(estudiante)
    
    @property
    def estudiantes(self):
        # Propiedad que devuelve la lista de estudiantes
        return self.__estudiantes
```

```python
class Estudiante:
    # Representa un estudiante con nombre y código
    def __init__(self, nombre, codigo):
        self.nombre = nombre
        self.codigo = codigo
    
    def mostrar_informacion(self):
        # Muestra el nombre y código del estudiante
        print(f"Nombre: {self.nombre}, Código: {self.codigo}")
```

```python
# Crear una universidad
uni = Universidad("Universidad Nacional")

# Crear una facultad
facultad = Facultad("Ingeniería")

# Crear al menos dos estudiantes
est1 = Estudiante("Ana López", "2024001")
est2 = Estudiante("Carlos Ruiz", "2024002")

# Agregar los estudiantes a la facultad
facultad.agregar_estudiante(est1)
facultad.agregar_estudiante(est2)

# Mostrar la información de cada estudiante usando el método de la clase Estudiante
for est in facultad.estudiantes:
    est.mostrar_informacion()
```
