## 1. Multiplicación rusa

Escribe un programa que multiplique dos números naturales mediante el algoritmo de la multiplicación rusa. Este algoritmo multiplica repetitivamente por dos el multiplicando y divide por dos el multiplicador, hasta que el multiplicador toma el valor 1. La suma todos los multiplicandos correspondientes a multiplicadores impares es el producto de los números originales.
Por ejemplo: $37 \times 12 = 12 + 48 + 384 = 444$, como se observa en la siguiente tabla:

| Multiplicador | Multiplicando |
| ------------- | ------------- |
|      37       |    **12**     |
|      18       |      24       |
|       9       |    **48**     |
|       4       |      96       |
|       2       |     192       |
|       1       |   **384**     |



````python
def multiplicacion_rusa(multiplicador, multiplicando):
    # Inicializar el resultado
    resultado = 0
    
    # Mostrar cabecera de la tabla
    print("Multiplicador | Multiplicando")
    print("-------------|-------------")
    
    # Bucle mientras el multiplicador sea mayor o igual a 1
    while multiplicador >= 1:
        # Si el multiplicador es impar, se suma el multiplicando y se marca en la tabla
        if multiplicador % 2 == 1:
            print(f"     {multiplicador}      |     {multiplicando}      *")
            resultado += multiplicando
        else:
            print(f"     {multiplicador}      |     {multiplicando}")
        
        # Actualizar: dividir multiplicador entre 2 (entero) y duplicar multiplicando
        multiplicador //= 2
        multiplicando *= 2
    
    # Mostrar el resultado final
    print(f"\nProducto = {resultado}")
    return resultado

# Ejemplo de uso
a = 37
b = 12
multiplicacion_rusa(a, b)
````

## 2. Suma de las cifras de un número

Escribe un programa que sume las cifras de un número. Por ejemplo, para $5127$ debe devolver $15$, ya que $5 + 1 + 2  +7 = 15$.

Sugerencia: dado un número, puedes utilizar los operadores `%` y `//` para obtener  su última cifra y todas sus cifras menos la última.

````python
def suma_cifras(numero):
    # Inicializar la suma
    suma = 0
    
    # Tomar el valor absoluto para manejar números negativos
    numero = abs(numero)
    
    # Mientras queden cifras en el número
    while numero > 0:
        # Obtener la última cifra usando módulo 10
        cifra = numero % 10
        # Sumar la cifra al total
        suma += cifra
        # Eliminar la última cifra usando división entera entre 10
        numero //= 10
    
    return suma

# Ejemplo de uso
n = 5127
resultado = suma_cifras(n)
print(f"La suma de las cifras de {n} es {resultado}")  # Salida: 15
````

## 3. Fizz Buzz

Fizz Buzz es un juego para enseñar a los niños a dividir mentalmente. En clase los niños van contando por turnos desde 1 (o desde otro número inicial). Si el número que le toca en su turno es divisible por 3, el niño debe decir **Fizz**; si es divisible por 5, debe decir **Buzz**; si es divisible por 3 y por 5, debe decir **Fizz Buzz**; si no es divisible ni por 3 ni por 5, debe decir el número que le toque. Por ejemplo, la secuencia para los números desde el 1 hasta el 36 es la siguiente: 

```
1, 2, Fizz, 4, Buzz, Fizz, 7, 8, Fizz, Buzz, 11, Fizz, 13, 14, Fizz Buzz, 16, 17, Fizz, 19, Buzz, Fizz, 22, 23, Fizz, Buzz, 26, Fizz, 28, 29, Fizz Buzz, 31, 32, Fizz, 34, Buzz, Fizz, ...
```

Escribe un programa  que ayude a los niños a contar desde un número inicial dado hasta que hayan dicho un número de veces **Fizz**, **Buzz** o **FIzz Buzz**.

````python
def fizz_buzz(inicio, fin):
    for i in range(inicio, fin+1):
        if i % 3 == 0 and i % 5 == 0:
            print("Fizz Buzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

fizz_buzz(5,15)
````

## 4. Cambio de base a binario

Escribe un programa que dado un entero en notación decimal escriba en pantalla su representación en binario.

````python
def decimal_a_binario(numero):
    # Manejar el caso especial del cero
    if numero == 0:
        return "0"
    
    # Tomar valor absoluto para números negativos
    # (mostrar el binario del valor absoluto con signo negativo)
    es_negativo = numero < 0
    num = abs(numero)
    
    # Lista para almacenar los restos (bits)
    bits = []
    
    # Dividir sucesivamente entre 2 hasta que el cociente sea 0
    while num > 0:
        resto = num % 2          # Obtener el bit (0 o 1)
        bits.append(str(resto))  # Guardar el resto
        num //= 2                # Actualizar el cociente
    
    # Los bits están en orden inverso (último resto es el bit más significativo)
    binario = ''.join(reversed(bits))
    
    # Agregar el signo negativo si corresponde
    if es_negativo:
        binario = '-' + binario
    
    return binario

# Solicitar el número al usuario
numero = int(input("Introduce un número entero en decimal: "))
resultado = decimal_a_binario(numero)
print(f"El número {numero} en binario es: {resultado}")
````

## 5. Cambio de base a decimal

Escribe un programa que dado un entero en binario lo pase a decimal, escribiendo su representación en pantalla.

````python
def binario_a_decimal(binario_str):
    # Inicializar el resultado
    decimal = 0
    
    # Recorrer cada dígito de izquierda a derecha
    for digito in binario_str:
        # Multiplicar el resultado actual por 2 y sumar el valor del dígito
        decimal = decimal * 2 + (1 if digito == '1' else 0)
    
    return decimal

# Solicitar el número binario al usuario (como cadena)
binario = input("Introduce un número en binario: ")
resultado = binario_a_decimal(binario)
print(f"El número binario {binario} en decimal es: {resultado}")
````
