### 1. Class inheritance:

```python
# ADD CODE. Define a Pet class. The constructor requires the name of the pet. 
# A class variable "counter" will be defined. It will be incremented in the constructor method and decremented in
# the destructor method.

class Pet:
    counter = 0

    def __init__(self, name):
        self.name = name
        Pet.counter += 1

    def __del__(self):
        Pet.counter -= 1
```

Cat and Dog inherit from the Pet class: 

```python
class Pet:
    counter = 0

    def __init__(self, name):
        self.name = name
        Pet.counter += 1
        print(f"Pet '{self.name}' created. Total pets: {Pet.counter}")

    def __del__(self):
        Pet.counter -= 1
        print(f"Pet '{self.name}' destroyed. Total pets: {Pet.counter}")

# Test
p1 = Pet("Fluffy")
p2 = Pet("Max")
del p1
# Output:
# Pet 'Fluffy' created. Total pets: 1
# Pet 'Max' created. Total pets: 2
# Pet 'Fluffy' destroyed. Total pets: 1
```

```python
class Pet:
    counter = 0
    def __init__(self, name):
        self.name = name
        Pet.counter += 1
    def __del__(self):
        Pet.counter -= 1

# Para diferenciar gato y perro, podrías crear subclases o simplemente usar Pet:
a_cat = Pet("Kitty")   # pero esto no especifica que sea un gato
# Mejor crear clases específicas:
class Cat(Pet):
    pass
class Dog(Pet):
    pass

a_cat = Cat("Kitty")
a_dog = Dog("Fido")
```

```python
# Definir el método purr para Cat (después de la herencia)
def purr(self):
    print(f"{self.name} says: Purrrr...")

Cat.purr = purr
```

```python
# Definir el método bark para Dog (después de la herencia)
def bark(self):
    print(f"{self.name} says: Woof!")

Dog.bark = bark
```

Shitzu and Beagle inherit from the Dog class:

```python
class Shitzu(Dog):
    def sleep(self):
        print(f"{self.name} sleeps 10 hours a day")
    
    def play(self):
        print(f"{self.name} plays with a toy")

class Beagle(Dog):
    def sleep(self):
        print(f"{self.name} sleeps 12 hours a day")
    
    def play(self):
        print(f"{self.name} plays in the yard")
```
