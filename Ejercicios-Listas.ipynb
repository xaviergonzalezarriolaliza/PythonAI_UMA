## 1. Media, máximo y mínimo

Usando iteración sobre listas, escribe un programa que calcule el máximo, el mínimo y la media aritmética de una lista de números. La respuesta se debe almacenar en una tupla con tres elementos. Por ejemplo, para la lista `[13, 8, 9, 45, 1, 0, 0, 6, 12]` se debe obtener la tupla `(45, 0, 10.44)`.

```python

# Define the list before the function
lista_numeros = [13, 8, 9, 45, 1, 0, 0, 6, 12]

def calcular_estadisticas():
    global lista_numeros
    maximo = lista_numeros[0]
    minimo = lista_numeros[0]
    suma = 0
    for numero in lista_numeros:
        if numero > maximo:
            maximo = numero
        if numero < minimo:
            minimo = numero
        suma += numero
    media = suma / len(lista_numeros)
    return (maximo, minimo, round(media, 2))

resultado = calcular_estadisticas()
print(resultado)  # (45, 0, 10.44)
```

## 2. Inserción en una lista ordenada

Escribe un programa que dadas una lista ordenada de menor a mayor y un valor, calcule la lista ordenada que resulta de añadir ese valor a la lista. Por ejemplo, para la lista `[1,3,5,7,9]` y el valor `4` debe calcular `[1,3,4,5,7,9]`. No puedes utilizar ninguna ordenación predefinida (`sort`, `sorted`, etc.).


```python
def insert_into_sorted(ordered_list, value):

    new_list = []
    inserted = False
    
    for item in ordered_list:
        # If value is smaller or equal to current item and not yet inserted,
        # insert value before this item (keeps stability: equal values go after existing ones? 
        # Actually this places value before first item >= value, so equal values go before existing equals.
        # To place after equals, change condition to item > value. Both are fine.)
        if not inserted and value <= item:
            new_list.append(value)
            inserted = True
        new_list.append(item)
    
    # If value is larger than all elements, append it at the end
    if not inserted:
        new_list.append(value)
    
    return new_list

# Example usage
original = [1, 3, 5, 7, 9]
value_to_insert = 4
result = insert_into_sorted(original, value_to_insert)
print(result)  # Output: [1, 3, 4, 5, 7, 9]

# Additional test cases
print(insert_into_sorted([], 10))           # [10]
print(insert_into_sorted([2, 4, 6], 1))     # [1, 2, 4, 6]
print(insert_into_sorted([2, 4, 6], 5))     # [2, 4, 5, 6]
print(insert_into_sorted([2, 4, 6], 6))     # [2, 4, 6, 6] or [2, 4, 6, 6] depending on condition
```

## 3. Elementos repetidos de una lista

Escribe un programa que dada una lista $x$ devuelva otra lista en la que figuren los elementos que aparecen repetidos en $x$. Por ejemplo, para la lista `[3, 1, 7, 3, 1, 1, 5, 7, 4, 2, 7]` debes obtener `[3, 1, 7]`.

```python
def find_repeated_elements(lst):
    # Count occurrences of each element
    count = {}
    for element in lst:
        if element in count:
            count[element] += 1
        else:
            count[element] = 1
    
    # Build list of repeated elements preserving first occurrence order
    repeated = []
    for element in lst:
        if count[element] > 1 and element not in repeated:
            repeated.append(element)
    
    return repeated

# Example usage
x = [3, 1, 7, 3, 1, 1, 5, 7, 4, 2, 7]
result = find_repeated_elements(x)
print(f"Repeated elements: {result}")  # Output: [3, 1, 7]
```

## 4. Eliminación de repetidos de una lista

Escribe un programa que elimine los elementos repetidos de una lista. De cada elemento repetido solo debe quedar su primera apacición. Por ejemplo, para la lista `[3, 1, 7, 3, 1, 1, 5, 7, 4, 2, 7]` debes obtener `[3, 1, 7, 5, 4, 2]`.

```python
def remove_duplicates(lst):
    # List to store unique elements in order
    unique = []
    # Set to keep track of elements we've already seen
    seen = set()
    
    # Iterate through the original list
    for element in lst:
        # If we haven't seen this element before, add it to the result
        if element not in seen:
            unique.append(element)
            seen.add(element)
    
    return unique

# Example usage
x = [3, 1, 7, 3, 1, 1, 5, 7, 4, 2, 7]
result = remove_duplicates(x)
print(f"Original list: {x} --->") 
print(f"List without duplicates: {result}")  # Output: [3, 1, 7, 5, 4, 2]
```

## 5. Elementos no repetidos en una lista

Escribe un programa que dada una lista cuente el número de elementos únicos (i.e. no repetidos) que aparecen en la misma. Por ejemplo, para la lista `[1,2,5,2,1,3]` debe mostrar `4`.

```python
import time

def remove_duplicates_new_list(lst):
    """Returns a new list with duplicates removed (preserves order)."""
    seen = set()
    unique = []
    for item in lst:
        if item not in seen:
            unique.append(item)
            seen.add(item)
    return unique

def remove_duplicates_inplace(lst):
    """Removes duplicates in‑place (modifies original list)."""
    seen = set()
    write_pos = 0
    for val in lst:
        if val not in seen:
            seen.add(val)
            lst[write_pos] = val
            write_pos += 1
    # Truncate the list to remove extra elements
    del lst[write_pos:]

# Create a test list with many duplicates
test_list = [i % 100 for i in range(1000000)]  # 10,000 elements, many repeats

# Measure new list approach
start = time.time()
new_result = remove_duplicates_new_list(test_list)
new_time = time.time() - start

# Measure in‑place approach
test_copy = test_list.copy()
start = time.time()
remove_duplicates_inplace(test_copy)
inplace_time = time.time() - start

print(f"New list approach: {new_time:.6f} seconds")
print(f"In‑place approach: {inplace_time:.6f} seconds")
print(f"Both produce the same result: {new_result == test_copy}")
print(f"Number of unique elements: {len(new_result)} of total {len(test_list)}")
```

Escribe un programa que dadas dos listas cuente el número de elementos que tienen en común, sin tener en cuenta los números repetidos. Por ejemplo, para las listas `[3,7,3,-1,2,3,7,2,15]` y `[15,2,-1,7,15,36,15]` debe mostrar `4`, pues el `-1`, `2`, `7` y `15` aparecen en ambas listas.

## 7. Mezcla de listas ordenadas

Escribe un programa que dadas dos listas ordenadas de menor a mayor calcule la lista ordenada con los elementos de ambas listas. Por ejemplo, para las listas `[1,3,5,7,9]` y `[2,4,5,6,8]` debe calcular `[1,2,3,4,5,5,6,7,8,9]`. No puedes utilizar ninguna ordenación predefinida (`sort`, `sorted`, etc.).

```python

def count_common_elements_manual(list1, list2):
    # Remove duplicates from list1
    distinct1 = []
    for item in list1:
        if item not in distinct1:
            distinct1.append(item)
    
    # Remove duplicates from list2
    distinct2 = []
    for item in list2:
        if item not in distinct2:
            distinct2.append(item)
    
    # Count how many elements of distinct1 appear in distinct2
    count = 0
    for item in distinct1:
        if item in distinct2:
            count += 1
    return count

# Define the two lists
a = [3, 7, 3, -1, 2, 3, 7, 2, 15]
b = [15, 2, -1, 7, 15, 36, 15]

# Test manual version
print(count_common_elements_manual(a, b))  # Output: 4
```
