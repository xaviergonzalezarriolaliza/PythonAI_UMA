## Necesidad de los diccionarios

Supongamos que queremos representar en Python el menú de una cafetería:

| Bebida         | Precio |
|----------------|--------|
| Americano      | 3000   |
| Iced Americano | 3500   |
| Cappuccino     | 4000   |
| Cafe Latte     | 4500   |
| Espresso       | 3600   |


Como vimos en la unidad anterior, debemos evitar el uso de variables disgregadas para representar cada valor (`cafe_1`, `precio_1`, ...). Es preferible utilizar una estructura de datos o colección que represente todo el menú de forma global. Una posibilidad es utilizar dos listas:

````python
cafes =   ["Americano", "Iced Americano", "Capuccino", "Cafe Latte", "Espresso"]
precios = [ 3000,        3500,             4000,        4500,         3600]
````

Ahora podemos buscar el precio de nuestra bebida en el menú:

````python
menu = {
    "Americano": 3000,
    "Iced Americano": 3500,
    "Cappuccino": 4000,
    "Cafe Latte": 4500,
    "Espresso": 3600
}

# Consultar precio de "Cappuccino" en O(1)
precio = menu["Cappuccino"]
print(precio)  # 4000

# Agregar nueva bebida
menu["Mocha"] = 4200

# Eliminar una bebida
del menu["Espresso"]

# Recorrer todo el menú
for bebida, precio in menu.items():
    print(f"{bebida}: ${precio}")
````

Esta solución presenta varios problemas:
- hay que mantener ambas listas _"sincronizadas"_: por ejemplo, si añadimos un café en la tercera posición de `cafes`, debemos añadir su precio exactamente en la tercera posición de `precios`
- hay que recorrer la lista `cafes` para saber el precio de un café: debemos localizar la posición del café y consultar dicha posición en la lista `precios`; esto tiene un coste lineal $O(n)$

Necesitamos una estructura de datos o colección que:
- nos libere de la necesidad de mantener la _"sincronización"_
- nos permita consultar de forma cómoda y eficiente el precio de un café

Tal estructura se llama **diccionario**, a veces también recibe los nombres de **aplicación**, **asociación** o **memoria asociativa**.

**Nota:** En algunos lenguajes (por ejemplo, Java) esta estructura se llama **map**.

## Definición de diccionario

Un **diccionario** es un conjunto de pares. Cada par tiene dos componentes: al primero lo llamamos **clave** y al segundo **valor**. Los pares suelen representarse como $k:v$, donde $k$ es la clave y $v$ es el valor.

No todo conjunto de pares $k:v$ es un diccionario válido: es necesario que **las claves no estén repetidas**. Por ejemplo, el siguiente conjunto de pares es un diccionario:

$$
\{ 1:10,\ 2:20,\ 3:30 \} 
$$

pero el siguiente conjunto de pares no es un dicionario, pues repite la clave $1$:

$$
\{ 1:10,\ 2:20,\ 1:40,\ 3:30 \} 
$$

Observa que los valores sí pueden estar repetidos:

$$
\{ 1:10,\ 2:20,\ 3:30,\ 4:10,\ 5:20,\ 6:30 \} 
$$

Si un diccionario contiene el par $k:v$, se dice que la clave $k$ tiene **asociado** el valor $v$. Por ejemplo, en el diccionario:

$$
\{ 1:10,\ 2:20,\ 3:30 \} 
$$

la clave $2$ tiene asociado el valor $20$. Como no puede haber claves repetidas, una clave puede tener asociado **como mucho un valor**. Esta característica es importante: dada una clave, podemos determinar el único valor que le corresponde.

Podemos representar el menú de la cafetería por el siguiente diccionario:

$$
\{ \mathrm{Americano}:3000,\ \mathrm{Iced\ Americano}:3500,\ 
   \mathrm{Cappuccino}:4000,\ \mathrm{Caffe\ Latte}:4500,\
   \mathrm{Espresso}:3600 \} 
$$

de manera que cada tipo de café (clave) tenga asociado su precio (valor).

## El tipo diccionario

El tipo diccionario es una colección o estructura de datos Python que permite asociar claves y valores, y acceder a los valores a través de las claves. Para determinar el valor que corresponde a una clave, es necesario que las claves no se repitan.

Un literal de diccionario tiene la siguiente sintaxis:

```python
      {k_0:v_0, k_1:v_1, k_2:v_2  ..., k_n:v_n}
```

Es decir, aparecen los pares `clave:valor` que contiene el diccionario, separados por comas y encerrados entre llaves.
Por ejemplo, podemos crear un diccionario con las edades de ciertas personas:

````python
# Diccionario con nombres como claves y edades como valores
edades = {
    "Ana": 25,
    "Luis": 30,
    "Carlos": 22,
    "Marta": 28
}

# Acceder a una edad mediante la clave
print(edades["Luis"])   # 30

# Agregar un nuevo par
edades["Elena"] = 27

# Modificar un valor existente
edades["Ana"] = 26

# Verificar si una clave existe
if "Pedro" in edades:
    print(edades["Pedro"])
else:
    print("Pedro no está en el diccionario")

# Eliminar una clave-valor
del edades["Carlos"]

# Recorrer claves y valores
for nombre, edad in edades.items():
    print(f"{nombre} tiene {edad} años")
````

Observa que podemos crear un diccionario vacío:

````python
# Ejemplo: contar frecuencias de palabras
frecuencias = {}          # empezamos vacío
palabras = ["hola", "mundo", "hola", "python"]

for palabra in palabras:
    if palabra in frecuencias:
        frecuencias[palabra] += 1
    else:
        frecuencias[palabra] = 1

print(frecuencias)   # {'hola': 2, 'mundo': 1, 'python': 1}
````

El tipo de las claves está limitado: las claves deben ser de tipos inmutables (`int` ,`float`, `str`,...). El tipo de los valores no está limitado:

````python
# Diccionario con distintos tipos de clave (inmutables) y valor (cualquier tipo)
ejemplo = {
    "nombre": "Ana",               # clave str, valor str
    42: "respuesta",               # clave int, valor str
    (1, 2): [3, 4, 5],             # clave tuple, valor lista
    3.14: {"pi": 3.1416},          # clave float, valor dict
    True: "verdadero"              # clave bool (subclase de int), valor str
}

# Imprimir todas las claves y valores
print("Claves y valores del diccionario:")
for clave, valor in ejemplo.items():
    print(f"Clave: {clave} (tipo: {type(clave).__name__}) -> Valor: {valor} (tipo: {type(valor).__name__})")
````

Observa que podemos usar listas como valores, pero no como claves (las listas son mutables):

````python
# Correcto: lista como valor
diccionario = {
    "frutas": ["manzana", "pera", "uva"],
    "verduras": ["lechuga", "tomate"]
}
print(diccionario["frutas"])      # ['manzana', 'pera', 'uva']

# Incorrecto: intentar usar lista como clave
try:
    d = {["manzana", "pera"]: "frutas"}
except TypeError as e:
    print(f"Error: {e}")   # TypeError: unhashable type: 'list'
````

El tipo de los diccionarios se llama `dict`, y no refleja ni los tipos de las claves ni los tipos de los valores:

````python
# Diccionario con claves y valores de diferentes tipos
d1 = {"nombre": "Ana", "edad": 25}
d2 = {1: "uno", 2: "dos"}
d3 = {(1,2): [3,4], 3.14: "pi"}

# El tipo de los tres es dict
print(type(d1))  # <class 'dict'>
print(type(d2))  # <class 'dict'>
print(type(d3))  # <class 'dict'>

# El tipo no refleja que d1 tiene claves str y valores str/int,
# ni que d2 tiene claves int, etc.
````

### Indexación de diccionarios

Una característica esencial de los diccionarios es que, dada una clave, podemos determinar el único valor que tiene asociado. Para ello, utilizamos la indexación por clave (similar a la indexación por posición en listas y otras secuencias):

```python
   diccionario[clave]
```

````python
precios = {"café": 1.20, "té": 1.00, "jugo": 1.50, "chocolate": 99.00}
print(precios["café"])   # 1.20
print(precios["té"])     # 1.00
print(precios["chocolate"])     # 99.00

print(precios.get("chocolate2", 0.0))   # 0.0 (no existe, devuelve el valor por defecto)
````

Si usamos una clave que no está presente en el `diccionario`; obtendremos un error de indexación `KeyError`:

````python
print(precios.get("piña"))      # None
print(precios.get("kiwi", 0.0)) # 0.0
if "zanahoria" in precios:
    print(precios["zanahoria"])
else:
    print("La clave no existe")
````

## Los diccionarios son mutables

Al igual que las listas, los diccionarios son mutables. Podemos añadir nuevos pares `k:v`, o bien modificar el valor asociado a una clave `k`. La sintaxis es similar a la que empleamos con las listas:

```python
   diccionario[clave] = valor
```

La anterior sentencia tiene uno de los siguientes efectos:
- si la `clave` no está presente en el `diccionario`, se añade el par `clave:valor`
- si la `clave` está presente en el `diccionario`, se modifica el `valor` asociado

````python
ventas = {"enero": 100, "febrero": 150}
print(ventas)  # {'enero': 100, 'febrero': 150}

# Modificar valor de una clave existente
ventas["febrero"] = 180
print(ventas)  # {'enero': 100, 'febrero': 180}

# Añadir un nuevo par (clave no existente)
ventas["marzo"] = 200
print(ventas)  # {'enero': 100, 'febrero': 180, 'marzo': 200}
````

## Eliminación de claves en diccionarios

Para eliminar un par `clave:valor` del diccionario, debemos utilizar la sentencia `del`:

```python
   del diccionario[clave]
```

Si la `clave` está en el `diccionario`, se elimina el par `clave:valor`, de lo contrario obtenemos un `KeyError`:

````python
inventario = {"manzanas": 10, "peras": 5, "naranjas": 8}
del inventario["peras"]
print(inventario)  # {'manzanas': 10, 'naranjas': 8}

# Si intentamos eliminar una clave inexistente:
if "uvas" in inventario:
    del inventario["uvas"]

else:
    print("La clave 'uvas' no existe, no se puede eliminar")
````

**Nota:** La sentencia `del` (de _delete_) puede utilizarse también para eliminar elementos de listas (por posición):

````python
lista = [10, 20, 30, 40]
del lista[2]        # elimina el elemento en índice 2 (el 30)
print(lista)        # [10, 20, 40]

del lista[1:3]      # elimina un rango
print(lista)        # [10]
````

De hecho, la sentencia `del` puede incluso borrar variables completas:

````python
x = 42
print(x)   # 42
del x
print(x)   # NameError: name 'x' is not defined
````

## Operadores y funciones sobre diccionarios

Los diccionarios soportan lo siguientes operadores:

| Operador     | Significado   |
|--------------|---------------|
|    `d[k]`    | indexación    |
|  `k in s`    | pertenencia   |
| `k not in s` | no pertenencia|
|  `d1 == d2`  | igualdad      |
|  `d1 != d2`  | desigualdad   |

Además, los diccionarios soportan la función `len()`.

````python
d = {"a": 1, "b": 2, "c": 3}

# Indexación
print(d["b"])          # 2

# Pertenencia
print("a" in d)        # True
print("z" not in d)    # True

# Igualdad
d2 = {"a": 1, "b": 2, "c": 3}
print(d == d2)         # True
print(d != d2)         # False

# Longitud
print(len(d))          # 3
````

Al comparar los diccionarios, el orden en que aparecen los pares no es relevante (son un _conjunto_ de pares):

````python
if "eva" in edades:
    print(edades["eva"])
else:
    print("Eva no está en el diccionario")

d1 = { 1: "uno", 2: "dos", 3: "tres" }
d2  = { 3: "tres", 2: "dos", 1: "uno" }
print(d1 == d2)   # True, el orden no importa en los diccionarios

print(d1 != d2)   # False, son iguales   
````

## Métodos sobre diccionarios

Los diccionarios soportan, entre otros, los siguientes métodos:

| Método      | Significado                                               |
|-------------|---------------------------------------|
| `d.clear()` | borra todos los pares del diccionario |
| `d.pop(k)`  | elimina la clave `k` de `d`       |


````python
d = {"a": 1, "b": 2, "c": 3}
valor = d.pop("b")
print(valor)  # 2
print(d)      # {'a': 1, 'c': 3}

# Con valor por defecto (evita KeyError)
valor = d.pop("z", "no existe")
print(valor)  # "no existe"
print(d)     # sigue siendo {'a': 1, 'c': 3}

d = {"a": 1, "b": 2, "c": 3}
d.clear()
print(d)   # {}

edades.pop("eva", -1)   # devuelve -1 si "eva" no existe
````

Observa que podemos eliminar una clave mediante `del` y `pop`. ¿Qué diferencia hay?

````python
d = {"x": 10, "y": 20}

# Usando del
del d["x"]          # elimina, no devuelve nada
print(d)            # {'y': 20}
# del d["z"]        # KeyError: 'z'

# Usando pop
valor = d.pop("y")  # devuelve 20 y elimina
print(valor)        # 20
print(d)            # {}

# pop con valor por defecto
valor2 = d.pop("z", 99)   # no hay error, devuelve 99
print(valor2)       # 99
````

## Iteración básica sobre diccionarios

Podemos usar un diccionario como fuente de datos en un bucle `for`. En tal caso, la variable de control tomará sucesivamente los valores de las **claves** del diccionario:

````python
edades = {
    "Ana": 25,
    "Luis": 30,
    "Carlos": 22,
    "Marta": 28
}
for nombre, edad in edades.items():
    print(f"{nombre} tiene {edad} años")
````

Usando iteración sobre diccionarios, podemos construir una lista de pares con la edad y el nombre de cada persona, ordenada por edades:

````python
# Reconstruir el diccionario edades
edades = {"Ana": 25, "Luis": 30, "Carlos": 22, "Marta": 28}

# Iteración básica: la variable toma las claves
for persona in edades:
    print(persona)           # Imprime Ana, Luis, Carlos, Marta (las claves)
    print(edades[persona])   # Imprime 25, 30, 22, 28 (los valores)
print();
for nombre, edad in edades.items():
    print(f"{nombre} tiene {edad} años")

# edad media
suma_edades = 0
for edad in edades.values():
    suma_edades += edad
media = suma_edades / len(edades)
print()
print(f"Edad media: {media:.2f}")
print()

# Ordenar por clave (nombre) y mostrar
print("Ordenado por nombre:")
for nombre in sorted(edades):
    print(f"{nombre}: {edades[nombre]}")

# Ordenar por valor (edad) y mostrar
print("\nOrdenado por edad:")
for nombre, edad in sorted(edades.items(), key=lambda item: item[1]):
    print(f"{nombre}: {edad}")
````

## El método `format`

Ya hemos visto que el método `format` permite asociar un formato a cada valor que queremos mostrar por pantalla:

````python
# Formato de números
print("Media: {:.2f}".format(10.4444))   # Media: 10.44
print("Entero: {:d}".format(42))         # Entero: 42
print("Porcentaje: {:.1%}".format(0.75)) # Porcentaje: 75.0%

# Múltiples valores
print("{} tiene {} años".format("Ana", 25))  # Ana tiene 25 años

# Con índices explícitos
print("{1} tiene {0} años".format(25, "Ana")) # Ana tiene 25 años

# Alineación
print("{:<10} {:>5}".format("Nombre", "Edad"))  # alineación izquierda/derecha
````

Hasta ahora hemos emparejado cada especificación de formato `{}` con el correspondiente dato facilitado en los argumentos de `format`:

````python
"{1:>10} tiene {0:.2f} años".format(25.567, "Ana")  # "       Ana tiene 25.57 años"
````

Es posible indicar en cada especificación de formato a qué argumento se refiere; basta indicar la posición del argumento (se numeran desde cero):

````python
"{1} tiene {0} años".format(25, "Ana")   # "Ana tiene 25 años"
````

Este índice de argumento se puede combinar con la especificación de formato propiamente dicha:

````python
# Primer argumento (índice 0) con 2 decimales, segundo (índice 1) con alineación a la derecha
resultado = "{0:.2f} - {1:>10}".format(3.1416, "Ana")
print(resultado)   # "3.14 -        Ana"
"{nombre:>10} tiene {edad:.1f} años".format(nombre="Ana", edad=25.567)
# "       Ana tiene 25.6 años"
````

## Solución del ejercicio de pair programming (147)

Como continuación del ejercicio anterior:
- Crea una *lista* llamada `dict_keys` que contenga las claves del diccionario y muestra la lista y su tipo.
- Define una *función* `is_fruit` que reciba el nombre de una fruta y verifique si está en el diccionario.
- *Llama* a la función con los valores `"apple"` y `"mango"` para comprobar su funcionamiento.

````python
# Diccionario de frutas y precios (del ejercicio anterior)
fruits_dic = {
    "apple": 5000,
    "banana": 4000,
    "grape": 5300,
    "melon": 6500
}

# 1. Crear lista dict_keys con las claves del diccionario
dict_keys = list(fruits_dic.keys())
print("Lista de claves:", dict_keys)
print("Tipo de dict_keys:", type(dict_keys))

# 2. Definir función is_fruit
def is_fruit(nombre):
    """Verifica si el nombre de la fruta está en el diccionario."""
    return nombre in fruits_dic

# 3. Llamar a la función con "apple" y "mango"
print(f"¿'apple' es una fruta de la lista? {is_fruit('apple')}")   # True
print(f"¿'mango' es una fruta de la lista? {is_fruit('mango')}")   # False
````

## Solución del mission problem (95)

Una cafetería ofrece las siguientes bebidas:

| Bebida         | Precio |
|----------------|--------|
| Americano      | 3000   |
| Iced Americano | 3500   |
| Cappuccino     | 4000   |
| Cafe Latte     | 4500   |
| Espresso       | 3600   |

Escribe un programa que muestre el menún y solicite al usuario la bebida que desea consumir. Si la bebida está en venta, se mostrará su precio. Si la bebida no está en venta, se informará al usuario de su error, y se volverá a mostrar el menú y a solicitar la bebida a consumir.

```
Welcome to David's cafeteria

Cafe      	Price
Americano 	3000
Iced Americano	3500
Capuccino 	4000
Cafe Latte	4500
Espresso  	3600

please, enter your choice:  carajillo

Sorry, we do not serve carajillo

Cafe      	Price
Americano 	3000
Iced Americano	3500
Capuccino 	4000
Cafe Latte	4500
Espresso  	3600

please, enter your choice:  Espresso

You selected Espresso Good choice!
Pease insert 3600 wons and enjoy
```

````python
# Diccionario con el menú de bebidas y precios
menu = {
    "Americano": 3000,
    "Iced Americano": 3500,
    "Cappuccino": 4000,
    "Cafe Latte": 4500,
    "Espresso": 3600
}

# Función para mostrar el menú
def mostrar_menu():
    print("\nCafe\t\tPrice")
    for bebida, precio in menu.items():
        print(f"{bebida}\t\t{precio}")

# Bucle principal
print("Welcome to David's cafeteria")
while True:
    mostrar_menu()
    opcion = input("\nPlease, enter your choice: ").strip()
    
    # Verificar si la bebida está en el menú (case-insensitive)
    # Buscar comparando sin distinción de mayúsculas/minúsculas
    bebida_encontrada = None
    for bebida in menu:
        if bebida.lower() == opcion.lower():
            bebida_encontrada = bebida
            break
    
    if bebida_encontrada:
        precio = menu[bebida_encontrada]
        print(f"\nYou selected {bebida_encontrada}. Good choice!")
        print(f"Please insert {precio} wons and enjoy.")
        break
    else:
        print(f"\nSorry, we do not serve {opcion}")
````

## Ejercicio extra

Dada una cadena, calcula una tabla de frecuencias absolutas de sus letras; es decir, una tabla que indique cuántas veces aparece cada letra en la cadena. Ten en cuenta que para calcular las frecuencias absolutas no tiene sentido distinguir entre mayúsculas y minúsculas.

**Sugerencia:** Utiliza un diccionario para representar la tabla de frecuencias. Las claves del diccionario serán las letras y los valores un entero positivo con el número de apariciones (por ejemplo, `"a":5` significa que la `a` aparece 5 veces en la cadena).

````python
import unicodedata

def normalizar(caracter):
    """Convierte una letra acentuada en su versión sin acento."""
    # Descompone el carácter (ej. 'á' -> 'a' + '́')
    descompuesto = unicodedata.normalize('NFD', caracter)
    # Toma solo la primera parte (la letra base) y la devuelve
    return descompuesto[0]

def frecuencia_letras(cadena):
    # Convertir a minúsculas para ignorar mayúsculas
    cadena = cadena.lower()
    frecuencias = {}
    
    for caracter in cadena:
        # Primero normalizar (quitar acentos)
        caracter_norm = normalizar(caracter)
        # Solo considerar si es una letra (después de normalizar)
        if caracter_norm.isalpha():
            # Incrementar contador usando get
            frecuencias[caracter_norm] = frecuencias.get(caracter_norm, 0) + 1
    
    return frecuencias

# Ejemplo de uso
texto = "Hola, cómo estás? ¡Qué bien!"
resultado = frecuencia_letras(texto)

# Ordenar: primero por frecuencia descendente, luego alfabéticamente
items_ordenados = sorted(resultado.items(), key=lambda item: (-item[1], item[0]))

print("Frecuencia de letras (ignorando acentos y mayúsculas):")
for letra, count in items_ordenados:
    print(f"{letra}: {count}")
````

## Ejercicio extra: clase Frecuencias

**Descripción:**  
Transforma el ejercicio anterior (cálculo de frecuencias absolutas de caracteres en una cadena) en una clase llamada `Frecuencias`.

**Requisitos:**

- **Atributos:**
  - `__cadena`: cadena de texto (atributo privado).
  - `__frecuencias`: diccionario vacío inicialmente (atributo privado).

- **Métodos:**
  - `__init__(self, cadena: str)`
    - Constructor que recibe la cadena a analizar.
    - Inicializa el diccionario `frecuencias` vacío.
  - `calcular(self)`
    - Calcula las frecuencias absolutas de cada carácter.
    - Almacena los resultados en el diccionario `__frecuencias`.
  - `__str__(self)`
    - Devuelve una representación legible del diccionario de frecuencias.

````python
class Frecuencias:
    """Clase para calcular frecuencias absolutas de letras en una cadena."""

    def __init__(self, cadena: str):
        self.__cadena = cadena
        self.__frecuencias = {}

    def calcular(self):
        texto = self.__cadena.lower()
        self.__frecuencias = {}
        for caracter in texto:
            if caracter.isalpha():
                self.__frecuencias[caracter] = self.__frecuencias.get(caracter, 0) + 1

    def __str__(self):
        if not self.__frecuencias:
            return "No se ha calculado ninguna frecuencia o la cadena no contiene letras."
        # Ordenar por frecuencia descendente, luego alfabéticamente
        items_ordenados = sorted(self.__frecuencias.items(),
                                 key=lambda item: (-item[1], item[0]))
        return "\n".join(f"{letra}: {count}" for letra, count in items_ordenados)


# Texto de prueba más largo (fragmento de Hamlet)
texto_largo = (
    "To be or not to be, that is the question: "
    "whether 'tis nobler in the mind to suffer "
    "the slings and arrows of outrageous fortune, "
    "or to take arms against a sea of troubles "
    "and by opposing end them."
)

frec = Frecuencias(texto_largo)
frec.calcular()
print("Frecuencias ordenadas por frecuencia (mayor a menor):")
print(frec)
````

## Ejercicio extra: Gestión de productos en una tienda

**Descripción:**  
Crea una clase llamada `Tienda` que permita gestionar un inventario de productos usando un diccionario.

**Requisitos:**

- **Atributos:**
  - `__nombre`: nombre de la tienda (privado)
  - `__inventario`: diccionario vacío inicialmente (privado)

- **Métodos:**
  - `__init__(self, nombre: str)`
    - Constructor que recibe el nombre de la tienda
    - Inicializa el diccionario `inventario` vacío
  - `agregar_producto(self, producto: str, cantidad: int)`
    - Agrega un producto al inventario o actualiza su cantidad si ya existe
  - `vender_producto(self, producto: str, cantidad: int) -> Bool`
    - Reduce la cantidad de un producto en el inventario si hay suficiente stock
    - Devuelve `True` si la venta fue exitosa, `False` si no hay suficiente stock o el producto no existe
  - `__str__(self) -> str`
    - Devuelve una representación legible del inventario

````python
class Tienda:
    """Gestiona el inventario de una tienda mediante un diccionario."""

    def __init__(self, nombre: str):
        """Constructor: almacena el nombre de la tienda e inicializa el inventario vacío."""
        self.__nombre = nombre
        self.__inventario = {}

    def agregar_producto(self, producto: str, cantidad: int):
        """Agrega o actualiza la cantidad de un producto en el inventario."""
        if cantidad <= 0:
            raise ValueError("La cantidad debe ser positiva.")
        if producto in self.__inventario:
            self.__inventario[producto] += cantidad
        else:
            self.__inventario[producto] = cantidad

    def vender_producto(self, producto: str, cantidad: int) -> bool:
        """Vende una cantidad de producto si hay suficiente stock."""
        if producto not in self.__inventario:
            return False
        if self.__inventario[producto] < cantidad:
            return False
        # Realizar la venta
        self.__inventario[producto] -= cantidad
        # Opcional: eliminar producto si la cantidad llega a cero
        if self.__inventario[producto] == 0:
            del self.__inventario[producto]
        return True

    def __str__(self) -> str:
        """Devuelve una representación legible del inventario."""
        if not self.__inventario:
            return f"Tienda '{self.__nombre}': Inventario vacío."
        # Mostrar productos ordenados alfabéticamente
        items = sorted(self.__inventario.items())
        inventario_str = "\n".join(f"  {prod}: {cant}" for prod, cant in items)
        return f"Tienda '{self.__nombre}':\n{inventario_str}"


# Ejemplo de uso
if __name__ == "__main__":
    tienda = Tienda("Mi Supermercado")
    tienda.agregar_producto("manzanas", 10)
    tienda.agregar_producto("naranjas", 5)
    tienda.agregar_producto("manzanas", 3)   # Actualiza a 13
    print(tienda)

    # Intentar vender
    print("\nVender 7 manzanas:", tienda.vender_producto("manzanas", 7))
    print(tienda)
    print("Vender 10 manzanas:", tienda.vender_producto("manzanas", 10))  # False (solo quedan 6)
    print("Vender 6 manzanas:", tienda.vender_producto("manzanas", 6))    # True
    print(tienda)
````
