## La selección o condicional simple

En la unidad anterior estudiamos una nueva estructura de control: la **selección o condicional simple**.



Esta estructura de control nos permite indicar que un paso es opcional; puede que se ejecute o puede que no, dependiendo del valor de un expresión booleana llamada condición o guarda.

Recuerda que la condicional simple se expresa en Python mediante la sentencia `if`:



## Limitaciones de la sentencia `if`

La sentencia `if` no es suficiente para resolver de forma adecuada algunos problemas; por ejemplo, saludar de una u otra forma según la hora del día:

````python
hora = int(input("qué hora es: "))

if hora <= 12:            # condicional simple
    print("buenos días")  # paso opcional

if hora > 12:
    print("buenas tardes")
````

El problema de la solución anterior es que estamos comprobando **explícitamente** una condición (`hour <= 12`) y su contraria o negación (`hour >12`).

Debería haber una forma más simple y eficiente de resolver estos problemas.

## La selección o condicional alternativa

La estructura de control de **selección o condicional alternativa** permite escoger entre dos alternativas **excluyentes** sin necesidad de indicar explícitamente las condiciones o guardas de cada una de ellas:




Observa que la condición o guarda se indica una sola vez. Si es cierta, se escoge una de las acciones; de lo contrario, se escoge la otra acción. Por eso decimos que las acciones son alternativas: o se ejecuta una o se ejecuta la otra.

## La sentencia `if else`

En Python la selección múltiple se expresa mediante la sentencia `if else`.

La sintaxis de la sentencia `if else` es una extensión de la sintaxis de la sentencia `if`:

```python
if condición:
    bloque_si_condición_es_cierta
else:
    bloque_si_condición_es_falsa
```

simplemente añadimos un bloque `else` a la sentencia `if` original.

La sentencia `if else` permite prescindir de la comprobación explícita de la condición negada; es decir, permite elegir entre dos opciones o alternativas excluyentes.


Ahora podemos revisar y simplificar nuestro programa de saludos:

````python
hora = int(input("qué hora es: "))

if hora <= 12:            # condicional simple
    print("buenos días")  # paso opcional

if hora > 12:
    print("buenas tardes")
````

````python
hora = int(input("qué hora es: "))

if hora <= 12:            # condicional alternativa
    print("buenos días")  # paso alternativo
else:
    print("buenas tardes") # paso alternativo
````

### Momento zen

Recuerda, hemos reemplazado dos sentencias `if` consecutivas con una condición y su contraria:
```python
if condición:
    bloque_si_condición_es_cierta

if not condición:
    bloque_si_condición_es_falsa
```
por una sentencia `if else` donde la condición aparece solo una vez:
```python
if condición:
    bloque_si_condición_es_cierta
else:
    bloque_si_condición_es_falsa
```

Es obvio que la segunda solución es más simple, legible y elegante. Pero ¿estos dos programas serán siempre equivalentes? ¿Qué piensas?

````python
hora = int(input("qué hora es: "))

if hora <= 12:            # condicional alternativa
    print("buenos días")  # paso alternativo
    hora = hora + 1
else:
    print("buenas tardes") # paso alternativo
````

````python
hora = int(input("qué hora es: "))

if hora <= 12:            # condicional simple
    print("buenos días")  # paso opcional
    hora = hora + 1

if hora > 12:
    print("buenas tardes")
````

## Control anidado

Hemos visto que dentro de una sentencia de control pueden aparecer otras sentencias que forman un bloque; es decir, una secuencia de sentencias que se ejecutan secuencialmente.

Por ejemplo, en este programa el bloque del `if` está compuesto por varias sentencias:

````python
x = int(input("dame un número natural x"))
y = int(input("dame un número natural y"))

if x >= 0 and y >= 0:
    print("x + y =", x + y)  # bloque con varias sentencias
    print("x - y =", x - y)
    print("x * y =", x * y)
    print("x / y =", x // y)
````

Es posible que dentro del bloque de una sentencia de control aparezca a su vez otra sentencia de control. A esto se le llama **control anidado**:

````python
x = int(input("dame un número natural x"))
y = int(input("dame un número natural y"))

if x >= 0 and y >= 0:
    print("x + y =", x + y)
    print("x - y =", x - y)
    print("x * y =", x * y)
    
    if y != 0:  # control anidado
        print("x / y =", x // y)
````

## Diseñando la lógica de nuestros programas

En estos momentos disponemos de varias herramientas para diseñar la lógica (flujo de control) de nuestros programas:

- secuencia
- selección simple
- selección alternativa
- control anidado

Debemos optar siempre por un diseño correcto y simple. Vamos a resolver un par de ejercicios utilizando las diferentes técnicas que hemos estudiado. 

**Ejercicio:** Vamos a escribir un programa Python que lea un entero por teclado e indique en un mensaje si el entero es:
- negativo,
- cero,
- positivo.

Observa que en esta ocasión tenemos tres alternativas excluyentes.

Podemos resolver nuestro problema mediante selección simple y sin control anidado:

````python
n = int(input("dame un número:"))

if n < 0:
    print("negativo")

if n == 0:
    print("cero")

if n > 0:
    print("positivo")
````

¿Qué ventajas e inconvenientes tiene la solución que usa selección simple?

También podemos resolver el problema utilizando selección alternativa y control anidado.

Por ejemplo, podemos anidar el control en el bloque del `if`:

````python
n = int(input("dame un número: "))

if n <= 0:
    # n <= 0
    if n < 0:
        # n < 0
        print("negativo")
    else:
        # n <= 0 and not (n < 0)
        # n <= 0 and n >= 0
        # n == 0
        print("cero")
else:
    # not (n <= 0)
    # n > 0
    print("positivo")
````

O bien podemos anidar el control en el bloque del `else`:

````python
n = int(input("dame un número: "))

if n < 0:
    print("negativo")
else:
    # qué es cierto aquí?
    # not (n < 0)
    # n >= 0
    if n == 0:
        print("cero")
    else:
        # n >= 0 and not (n == 0)
        # n >= 0 and n != 0
        # n > 0
        print("positivo")
````

¿Qué ventajas e inconvenientes tiene la solución que usa selección alternativa?

**Ejercicio:** Escribe un programa Python que lea una nota numérica $n$ por teclado (supón que está entre $0$ y $10$) e indique qué calificación se ha obtenido según la siguiente escala:

| Nota $n$ | Calificación |
|-|-|
| $0 \leq n < 5$ | suspenso |
| $5 \leq n < 7$ | aprobado |
| $7 \leq n < 9$ | notable |
| $9 \leq n \leq 10$ | sobresaliente |

Tienes que escribir dos soluciones:
- usando solo condicional simple (sentencia `if`) sin control anidado
- usando solo selección alternativa (sentencia `if else`) y control anidado

````python
n = float(input("Añade la nota: "))

if 0 <= n < 5:
    print("suspenso")

if 5 <= n < 7:
    print("aprobado")

if 7 <= n < 9:
    print("notable")

if 9 <= n <= 10:
    print("sobresaliente")
````

````python
n = float(input("Añade la nota: "))

if n < 5:
    # n <= 5
    print("suspenso")
else:
    # not (n < 5) 
    # n >= 5
    if n < 7:
        # 5 <= n < 7 
        print("aprobado")
    else:
        # n >= 5 and not (n < 7)
        # n >= 5 and n >= 7
        # n >= 7
        if n < 9:
            #  n >= 7 and n < 9
            #   7 <= n < 9
            print("notable")
        else:
            # n >= 7 and not (n < 9)
            # n >= 7 and n >= 9
            # n >= 9
            print("sobresaliente")
````

````python
n = float(input("Añade la nota: "))

if 0 <= n < 7:
    if n < 5:
        print("suspenso")
    else:
        print("aprobado")
else:
    if n < 9:
        print("notable")
    else:
        print("sobresaliente")
````

## La selección múltiple

Situaciones como la anterior son muy comunes: lo que queremos es elegir una opción entre varias posibles que son excluyentes entre sí. Por ejemplo, clasificar un número según su signo (negativo, cero o positivo) o clasificar una nota (suspenso, aprobado, notable, sobresaliente).

Aunque es posible resolver estos problemas con las herramientas que hemos estudiado, la solución es más complicada de lo necesario. Lo que necesitamos es una nueva estructura de control que nos permite elegir una entre varias opciones. A esta estructura se le llama **selección o condicional múltiple**:



En Python, la selección múltiple se expresa mediante la sentencia `if elif else`, cuya sintaxis es:

```python
if condición_1:
    bloque_si_condición_1_es_cierta
elif condición_2:
    bloque_si_condición_2_es_cierta
elif condición_3:
    bloque_si_condición_3_es_cierta
    
 # tantas condiciones como sea necesario...

else:
    bloque_si_ninguna_condición_es_cierta
```
Esta nueva sentencia nos permite reescribir los programa anteriores de forma más clara y simple.

**Ejercicio:** Vamos a escribir un programa Python que lea un entero por teclado e indique en un mensaje si el entero es:

Utiliza selección múltiple.

````python
n = int(input("dame un número:"))

if n < 0:
    print("negativo")
elif n == 0:
    print("cero")
else:
    print("positivo")
````

## Sentencias condicionales y expresiones booleanas

Las sentencias condicionales, como todas las sentencias de control, dirigen el flujo del programa según el valor de las expresiones booleanas que actúan de guardas. Por ello, es muy importante entender bien las expresiones booleanas.

La siguiente expresión booleana es una solución al problema de los años bisiestos:

````python
year = 1800
````

````python
(year % 4 == 0 and year % 100 != 0) or year % 400 == 0
````

Aquí tenemos otra solución al mismo problema:

````python
year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)
````

Aunque las soluciones se parecen mucho, son ligeramente diferentes.

Las siguientes expresiones booleanas siguen el esquema de las soluciones anteriores y devuelven valores diferentes para los mismos valores de `a`, `b` y `c`:

````python
a = False
b = False
c = True
````

````python
(a and b) or c
````

````python
a and (b or c)
````

En el caso de los años bisiestos la diferencia no importa, porque las condiciones `a` (`year % 4 == 0`) y `c` (`year % 400 == 0`) no son independientes: si se cumple `c`, se cumple `a`. La segunda solución a los bisiestos parece mejor, pues detalla por separado la regla general del calendario juliano (múltiplos de 4) de la excepción del calendario gregoriano (años centenarios).

## Números aleatorios en Python

Python no sabe nada acerca de números aleatorios, como tampoco sabe de Aprendizaje Automático (_Machine Learning_).
Sin embargo, Python puede _aprender_ cosas nuevas: basta **importar** un módulo o biblioteca para que Python adquiera nuevas habilidades. Para importar un módulo usamos la sentencia `import`:

> `import` _nombre_del_modulo_

````python
import random
````

El módulo `random` le _enseña_ a Python cómo generar números aleatorios.

No es necesario saber cómo está hecho un módulo por dentro ni cómo funciona; basta saber usarlo, es decir, saber qué funciones le _enseña_ a Python. Para el módulo `random` nos interesan dos funciones:

| Función          | Significado                    |
|------------------|--------------------------------|
| `randint(l,u)`   | devuelve un aleatorio en [l,u] |
| `randrange(l,u)` | devuelve un aleatorio en [l,u) |


En general, para saber qué nos ofrece un módulo podemos consultar su documentación o usar la función `help()`.

````python
random.randint(1,6)
````

````python
random.randrange(1,7)
````

````python
for i in range(1,7):
    print(i)
````
